```python
"""
自定义异常捕获：
用途：借用python的异常处理机制，编写程序中断逻辑，继承现有的异常，并为现有的异常添加新的额外功能
方法：继承Exception类，并重写__init__方法，在__init__方法中添加新的功能
try:
    raise MyException
except MyException as e:
    print(e)
"""
```

```python
class MyException(Exception):
    def __init__(self,message):   #message是传入的参数
        self.name=message
    @ property
    def msg(self):
        return f"名字不允许被使用{self.name}"
```

```python
name = "jerry"
try:
    if name == "jerry":
        raise MyException(name)  #   raise关键字   抛出自定义异常
except MyException as e:         # 捕获自定义异常    as关键字，实例化
    print(e.msg)
```
