```python
#并行和并发区别
"""
并发：宏观上，多个任务一起执行，微观上，多个任务交替执行
并行：多个任务同时执行

并发：一个CPU核心交替运行多个任务
并行：多个CPU核心同时运行多个任务
"""
```

```python
"""
利用concurrent.futures模块实现并行计算
Executor对象并行任务的抽象类
由线程（不能跨CPU）和进程两种实现方式
Executor.submit()方法提交任务

Executor对象支持ThreadPoolExecutor,使用线程池实现并行计算
Executor对象支持ProcessPoolExecutor,使用进程池实现并行计算，多核CPU
"""
```
