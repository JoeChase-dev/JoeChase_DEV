```python
"""
魔术方法：
以__ 和 __ 开始结束，实现特殊功能
可以在原类型中添加自己的方法
"""
```

```python
"""
__init__()  称作初始化方法，属性和方法的初始化
类实例化时进行
还可以让类实例化时接收参数
"""
```

```python
class Klass(object):
    def __init__(self, name,age):  # 接收多少个参数，就写多少个name
        # 传递参数   自动执行
        self.name = name
        self.age = age
    def information(self):
        print("name:", self.name)
        print("age:", self.age)

obj = Klass('Wilson',18)
obj.information()
```
