```python
"""
类的装饰器
改变方法的调用方式和行为

"""
```

```python
#classmethod 实现实例的方法定义为类的方法，用于类的直接调用
#避免实例化，实现同一属性的统一处理
class Klass1:
    @classmethod
    def func(cls):
        print('classmethod')
Klass1.func()
#cls 表示当前操作的类，可以使用Klass.func()调用
```

```python
#staticmethod 不需要类的任何信息但又和类相关的一些方法，维护代码并保持代码公整
#不需要self cls
class Klass2:
    @staticmethod
    def func():
        print('staticmethod')
```

```python
#property 1 将方法装饰成属性     2 让属性像方法一样实现复杂逻辑
#将调用难度降低，将方法调用变为属性调用
class Klass3:
    @property
    def func(self):           # 第1步：定义"假属性"的获取逻辑
        print('property')     #      访问时打印
        return self.__varValue  #    返回私有变量

    @func.setter                #装饰已经变为属性的func函数  setter 实现属性设置逻辑，增加功能
    def func(self, value):    # 第2步：定义"假属性"的设置逻辑
        self.__varValue = value  #  存到私有变量

obj = Klass3()        # 创建对象
obj.func = '123'      # ← 触发 setter：把 '123' 存进 __varValue
print(obj.func)       # ← 触发 getter：打印'property'，返回 '123'
```
