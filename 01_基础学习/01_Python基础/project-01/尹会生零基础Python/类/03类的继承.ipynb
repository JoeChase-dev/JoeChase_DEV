```python
"""
可以快速上手，可以增加功能

父类和子类
在继承中，被继承的类称作父类，继承的类称作子类
列的继承指子类继承了父类的属性、方法
"""
```

```python
#增加功能时可以使用super()方法
class Father(object):
    def run(self):
        print('run father')

class Son(Father): # 继承父类(被继承的父类)
    def run(self):
        super().run() #继承父类的方法
        print("run")
```

```python
obj = Father()
obj.run()
```

```python
obj1 = Son()
obj1.run()
```

```python
#类的多继承
"""
菱形继承时，Python会按照C3算法按顺序遍历继承图
通过类对象名称.__mor__可以查看继承顺序
多重继承增加了继承的复杂度，应当减少多重继承的使用
"""
```
