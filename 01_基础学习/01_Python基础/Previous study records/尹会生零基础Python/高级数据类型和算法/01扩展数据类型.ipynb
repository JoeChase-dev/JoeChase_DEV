```python
"""
命名元组：
namedtuple()是命名元组的工厂函数    像函数一样使用
命名元组使用前需要导入模块： collections
定义了一个点的坐标，x,y
Point = namedtuple('Point', ['x', 'y'])
"""
```

```python
from collections import namedtuple
Point = namedtuple('Point', ['x', 'y'])
p1 = Point(11,y=12)
```

```python
p1.x is p1[0]
```

```python
p2 = Point(13,y=14)
p2 + p1
```

```python
class Point_2D(namedtuple('Point', ['x', 'y'])):
    def __add__(self,other):
        self.other_x,self.other_y = other
        return self.x+self.other_x,self.y+self.other_y
```

```python
p1 = Point_2D(11,y=12)
p2 = Point_2D(13,y=14)
p1 + p2    # 相当于p1.__add__(p2)
```

```python
"""
字典和列表子类化：
UserDict 字典子类化
UserList 列表子类化
可以通过继承 UserDict 和 UserList 来实现自定义的字典和列表
"""
```

```python
from collections import UserDict
class MyDict(UserDict):
    def __setitem__(self, key, value):
        if key in self.data.keys():    # data 是 UserDict 中的一个属性，用于存储字典中的元素，相当于原来的字典
            print(f"字典中已有该键，请勿重复添加{key}")
        else:
            self.data[str(key)] = value
```

```python
dict = MyDict()
```

```python
dict['a'] =1
```
