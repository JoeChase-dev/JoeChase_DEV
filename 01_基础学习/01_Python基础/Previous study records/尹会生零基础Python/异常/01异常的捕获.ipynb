```python
#异常的捕获
try:
    #可能会出错的代码
except #异常:
    #捕获指定异常后运行的代码
```

```python
1/0
```

```python
try:
    1/0
except ZeroDivisionError:
    print('除数不能为0')
```

```python
num = int(input('请输入一个数字:'))
num2= int(input('请输入一个数字:'))
try:
    num/num2
except Exception as e:#  捕获 e 是异常对象，将异常存在e中
    print("上一段程序有异常")
    print(e)
```

```python
"""
finally 代码块
try:
    可能会出错的代码
except:
    捕获异常后运行的代码
finally:
    无论是否有异常，都会运行的代码
"""
```
