```
"""1-100内的素数
说明：素数指的是只能被 1 和自身整除的正整数（不包括 1）。
"""
# 自我版
for i in range(2,101):
    is_prime = True
    for j in range(2,int(i**0.5)+1):# range的类型必须为整数，所以要强制转换为整数，i**0.5为浮点数
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i,end=' ')
```

```
"""
斐波那契数列
特点，前两个数是1，后面的数是前两个数的和
输出前二十个数
"""
#  a = 0 ,b = 1   实际是 a = (0,b=1)   Python不允许在赋值表达式的右边使用赋值表达式
a,b = 0,1
for _ in range(20):
    a,b = b,a+b
    print(a,end=' ')
```

```
"""
输出10000以内的水仙花数
说明：水仙花数是指一个 3 位数，它的每个位上的数字的 3次幂之和等于它本身。
N位非负整数，各个数字的N次方刚好等于该数本身
"""
for i in range(100,1000):
    low = i % 10
    middle = i // 10 % 10
    high = i //100
    if i == low**3 + middle**3 + high**3:
        print(i,end=' ')
```

```
"""
百鸡百钱
公鸡五元一只，母鸡三元一只，小鸡一元三只，用100元买100只鸡，问公鸡，母鸡，小鸡各有多少只？
"""
# 方法一  穷举法，暴力搜索法
for x in range(0,21):
    for y in range(0,34):
        for z in range(0,100,3):
            if 5*x+3*y+z//3 == 100 and x+y+z == 100:
                print(f'公鸡：{x}只，母鸡：{y}只，小鸡：{z}只')
# 方法二
for x in range(0,21):
    for y in range(0,34):
        z = 100 - x - y
        if 5*x+3*y+z//3 == 100 and z%3 == 0:
            print(f'公鸡：{x}只，母鸡：{y}只，小鸡：{z}只')
```

```
"""
CRAPS赌博游戏
"""
import random as rd

money = 1000
while money > 0:
    print(f'当前余额为：{money}')
    while True:
        debt = int(input('请输入赌注：'))
        if 0 < debt <= money:
            print(f'您本次下注了{debt}元')
            break
    first_point = rd.randint(1,7) + rd.randint(1,7)
    print(f'第一次点数为{first_point}')
    if first_point == 7 or first_point == 11:
       print(f'第一次点数为{first_point}，您赢了')
       money += debt
       print(f'当前余额为：{money}')
    elif first_point == 2 or first_point == 3 or first_point == 12:
        print('您输了')
        money -= debt
        print(f'当前余额为：{money}')
    else:
        while True:
            second_point = rd.randint(1,7) + rd.randint(1,7)
            print(f'第二次点数为{second_point}')
            if second_point == 7:
                print('您输了')
                money -= debt
                print(f'当前余额为：{money}')
                break
            elif second_point == first_point:
                print('您赢了')
                money += debt
                print(f'当前余额为：{money}')
                break
print('您已经破产，游戏结束')
```
