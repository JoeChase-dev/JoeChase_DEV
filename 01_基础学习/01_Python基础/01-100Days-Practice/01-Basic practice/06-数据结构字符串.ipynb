```python
"""
原始字符串
以r或R开头，表示原始字符串，不转义
"""
s = '\nasddgh\tasfgg\nasfg'
s1 = r'\nasddgh\tasfgg\nasfg'
print(s)
print(s1)
```

```python
"""
字符串的运算
"""
# 拼接和重复
s = 'hello '
s1 = 'Python'
s3 = '！'
s3 *= 3
print(s + s1 + s3)
print(s*3)
print(s1 + s3)


print('\n')


# 成员运算  in / not in
s1 = 'hello world'
s2 = 'world'
print(s2 in s1)  # True
print(s2 not in s1)  # False

print('\n')

# 切片
s1 = 'hello world'
# 字符串不可变类型，只能通过索引实现访问，不可以通过索引修改
```

```python
"""
字符串的遍历
"""
# 索引遍历
s1 = 'Hello Python!'
for i in range(len(s1)):
    print(s1[i], end=' ')
print('\n')
#
for i in s1:
    print(i, end=' ')
```

```python
"""
大小写操作
"""
s1 = 'hello python'
print(s1.capitalize())
print(s1.title())
print(s1.upper())
print(s1.lower())
print(s1)
```

```python
"""
查找
"""
s1 = 'hello,world'
print(s1.find('or'))
print(s1.find('or',9))  # 从9开始查找，找不到返回-1
print(s1.index('or'))
print(s1.index('or',9))  # 报错 ValueError: substring not found
# 逆向查找 rfind / rindex
```

```python
"""
性质判断
"""
s1 = 'hello,world!'
print(s1.startswith('h'))  # 以h开头，返回True，否则返回False
print(s1.endswith('!'))   # 以!结尾，返回True，否则返回False
s2 = 'abc123'
print(s2.isalnum())# 是否是字母和数字
print(s2.isalpha())# 是否是字母
print(s2.isdigit())# 是否是数字
```

```python
"""
格式化；
"""
s1 = 'hello,world'
print(s1.center(20,'*'))  # 居中，20个字符，不足补*
print(s1.ljust(20,'~'))  # 左对齐，20个字符，不足补~
print(s1.rjust(20,))  # 右对齐，20个字符，不足补空格
print('33'.zfill(5)) # 左补零
print('-33'.zfill(5))  #
```

```python
"""修剪操作
可以用来避免用户错输入空格
"""
s1 = '   hello,world    '
print(s1.strip())
s2 = '~~hello,world~~'
print(s2.strip('~'))
print(s2.lstrip('~'))
print(s2.rstrip('~'))
```

```python
"""
替换
"""
s = 'HELLO,WORLD'
print(s.replace('O','@'))
print(s.replace('LO','!!'))
```

```python
"""
拆分与合并
"""
s = 'i love you'
word = s.split()
print(word)
print('~'.join(word))
```
