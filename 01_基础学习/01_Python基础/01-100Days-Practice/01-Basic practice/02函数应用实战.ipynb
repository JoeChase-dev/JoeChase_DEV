```python
import random
import string
ALL_CHARS = string.ascii_letters + string.digits
# digits 代表0-9的数字所构成的字符串
# ascii_letters 代表a-zA-Z的字符串


def random_code(*,code_len=4):
# * 仅限关键字参数，必须使用关键字传参，不能使用位置传参
# * 目的：1.清晰性，调用时必须写参数名，代码容易读：灵活性：添加新的参数，不影响现在调用

    """
    生成指定长度的验证码
    :param code_len: 验证码的长度
    :return: 有大小写字母和数字组成的随机验证码
    """
    return ''.join(random.choices(ALL_CHARS,k=code_len))
# k 代表从ALL_CHARS中随机选择k个元素
# ''.join() 将列表中元素用指定字符' '串连接
# sample  随机抽样 ， 无放回抽样
# choices  随机抽样，可重复选中，k代表样本容量
```

```python
for _ in range(5):   # _ 约定俗成的变量名，表示不在乎，在循环体中不使用
    print(random_code())
```

```python
"""
判断素数：
大于一的正整数是不是质数，质数是只能被1和自身整除的数，大于1，是质数则在2-n-1之间都不能被整除
sqrt 开平方  之后也没有了"""
def is_prime(num:int)->bool:
    # num:int 代表num是int类型，->bool 代表返回值是bool类型
    """
    判断是否是素数
    :param num:
    :return:
    """

    for i in range(2,int(num**0.5)+1):# int表示强制转换，因为range()函数要求参数是int类型，num**0.5为浮点数
        if num % i == 0:
            return False
    return True

print(is_prime(10))
```

```python
"""
最大公约数和最小公倍数
最大公约数是同时整除x和y的最大数，如果x和y互质，则最大公约数为1
最小公倍数是同时被x和y整除的最小数，如果x和y互质，则最小公倍数为x*y
"""

def gcd(x:int,y:int):
    """
    最大公约数
    :param x:
    :param y:
    :return:
    """
    while y % x != 0:
        x,y = y % x,x # 欧几里得算法
        #如果有两个正整数  a和b，则它们的最大公约数等于a除以b的余r与b的最大公约数。
    return x
def lcm(x:int,y:int):
    """
    最小公倍数
    :param x:
    :param y:
    :return:
    """
    return x*y//gcd(x,y)  # 数学公式
print(gcd(10,20))
print(lcm(10,20))
```

```python
"""
数据统计，设计计算样本数据描述性统计信息的函数
"""
list1 = [1234,35,24,64,85,523,25,867,69]
def ptp(data:list):
    """
    极差（全距）
    :param data:
    :return:
    """
    return max(data) - min(data)
def mean(data:list):
    """
    算数平均
    :param data:
    :return:
    """
    return sum(data)/len(data)
def median(data:list):
    """
    中位数
    :param data:
    :return:
    """
    temp,size = sorted(data),len(data) # 长度比索引大一个数
    if size % 2 ==1:
        return temp[size//2]
    else:
        return mean(temp[size//2-1:size//2+1])
        # temp[size//2-1:size//2+1] 代表切片
```

```python
def var(data,ddof=1):
    """
    方差
    方差表示一组数据的离散程度(数据分散得有多开)
    方差大 → 数据很分散,差异大
    方差小 → 数据很集中,差异小
    总体方差	(n)	拥有全部数据时	ddof=0
    样本方差	(n-1)	只有一部分样本时	ddof=1
    记忆口诀:
    全部数据 → ddof=0(不减)
    部分样本 → ddof=1(减1)
    :param data:
    :param ddof:
    :return:
    """
    x_bar = mean(data)
    temp = [(num-x_bar)**2 for num in data]
    return sum(temp)/(len(data)-ddof)
```

```python
def std(data,ddof=1):
    """
    标准差
    :param data:
    :param ddof:
    :return:
    """
    return var(data,ddof)**0.5
```

```python
def cv(data,ddof=1):
    """
    变异系数
    :param data:
    :param ddof:
    :return:
    """
    return std(data,ddof)/mean(data)
```

```python
def describe(data):
    """
    描述性统计
    :param data:
    :return:
    """
    print(f'均值：{mean(data)}')
    print(f'中位数：{median(data)}')
    print(f'极差：{ptp(data)}')
    print(f'方差：{var(data)}')
    print(f'标准差：{std(data)}')
    print(f'变异系数：{cv(data)}')
```

```python
"""
双色球随机选号
"""
import random
RED_BALLS = [i for i in range(1,34)]
BLUE_BALLS = [i for i in range(1,17)]
def choose():
    """
    随机生成双色球号码
    :return:
    """
    selected_red = random.sample(RED_BALLS,6)
    selected_red.sort()
    selected_red.append(random.choice(BLUE_BALLS))
    return selected_red
def display(balls:list):
    """
    格式输出一组号码
    :param balls:
    :return:
    """
    for ball in balls[::-1]:
        print(f'\033[031m{ball:0>2d}\033[0m',end=' ')
        # \033[031m 是红色颜色代码，
    print(f'\033[034m{balls[-1]:0>2d}\033[0m,')
    # \033[034m 是蓝色颜色代码

n= int(input('生成几注号码：'))
for _ in range(n):
    display(choose())
```
