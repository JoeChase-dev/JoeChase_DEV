```python
"""
视图对象：
dict.keys() / dict.values() / dict.items() 视图对象详解

不可索引，不能切片，动态同步


"""
d = {"apple": 3, "banana": 5, "orange": 2}

# Python 2 时代：返回的是 列表
# Python 3 时代：返回的是 视图对象 (View Object)

keys = d.keys()      # dict_keys(['apple', 'banana', 'orange'])
values = d.values()  # dict_values([3, 5, 2])
items = d.items()    # dict_items([('apple',3), ('banana',5), ('orange',2)])

print(type(keys))    # <class 'dict_keys'>
print(type(values))  # <class 'dict_values'>
print(type(items))   # <class 'dict_items'>


# 就像"实时镜像"——视图始终反映字典的最新状态
#视图对象的"动态同步"特性
d = {"a": 1, "b": 2}

view = d.keys()
print(view)        # dict_keys(['a', 'b'])

# 字典发生变化后...
d["c"] = 3
print(view)        # dict_keys(['a', 'b', 'c'])  ← 自动更新了！

del d["a"]
print(view)        # dict_keys(['b', 'c'])       ← 又自动更新了！

"""
什么时候需要转成列表？
当你需要按索引访问或多次操作时
"""
d = {"apple": 3, "banana": 5}

# ❌ 这样会报错
keys = d.keys()
print(keys[0])     # TypeError: 'dict_keys' object is not subscriptable

# ✅ 需要转为列表
keys_list = list(d.keys())
print(keys_list[0])  # 'apple'
print(keys_list[1:]) # ['banana']


# 简洁写法
best_student = max(scores, key=scores.get)  # 一行搞定！
# 但当我们想要自定义比较规则时，就用 key= 参数：
# key=scores.get 的意思是：对每个键，用 scores.get(键) 得到分数，按分数比较大小

"""

┌─────────────────────────────────────────────┐
│         dict 的三个视图方法                    │
├──────────┬────────────┬─────────────────────┤
│ 方法      │ 返回类型     │ 示例                 │
├──────────┼────────────┼─────────────────────┤
│ .keys()  │ dict_keys  │ 所有的键             │
│ .values()│ dict_values│ 所有的值             │
│ .items() │ dict_items │ 所有的(键,值)元组      │
└──────────┴────────────┴─────────────────────┘

需要索引/切片？→ 用 list() 包起来！
list(d.keys())    → ['a', 'b', 'c']  ✅
list(d.values())  → [1, 2, 3]        ✅
list(d.items())   → [('a',1), ...]   ✅
"""
```

```python
"""
else 语句的独特之处：


for ... else
├── 循环正常跑完（没被 break）→ 执行 else
└── 被 break 打断              → 跳过 else

while ... else
├── 循环条件变为 False 正常退出 → 执行 else
└── 被 break 打断             → 跳过 else

"""
# 1️⃣ if...else   （所有语言都有）
if condition:
    ...
else:
    ...

# 2️⃣ for...else    （Python 特色！⭐）
for i in range(10):
    ...
else:
    ...

# 3️⃣ while...else  （Python 特色！⭐）
while condition:
    ...
else:
    ...
"""
测试代码：
"""
# 测试1：有 break
print("=== 测试1: 有 break ===")
for i in range(3):
    if i == 1:
        break
    print(i)
else:
    print("else执行了!")
print()

# 测试2：无 break
print("=== 测试2: 无 break ===")
for i in range(3):
    print(i)
else:
    print("else执行了!")
```
