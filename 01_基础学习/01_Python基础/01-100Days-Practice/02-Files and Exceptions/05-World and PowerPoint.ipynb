```python
"""
可以使用名为python-docx 的三方库来操作 Word，
可以使用名为python-pptx的三方库来生成 PowerPoint。
"""
# 从第三方库中导入Document类
from docx import Document
# 导入单位类，Cm 表示厘米(设置宽度)，Pt 表示磅/字号
from docx.shared import Cm,Pt

from docx.document import Document as Doc

# 创建DOC对象，实例化
document = Document() #type: Doc
# 添加大标题 0 = 一级标题  最大       arg1: 标题内容  arg2: 标题级别
document.add_heading('快乐学Python',0)


# 添加段落,段落中添加文本片段Run
p = document.add_paragraph('Python是一门非常流行的语言，它')
# run 是文本片段，可以独立设置格式
# 添加文本片段
run = p.add_run('简单')
run.bold = True # 加粗
run.font.size = Pt(18) # 字号
p.add_run('而且')
run = p.add_run('优雅')
run.font.size = Pt(18)
run.underline = True
p.add_run('。')

# 添加一级标题
document.add_heading('Heading,level 1',level = 1)


# 添加带样式的段落
document.add_paragraph('Intense quote',style = 'Intense Quote')
# 添加无序列表  圆点
document.add_paragraph('first item in unordered list',style = 'List Bullet')
document.add_paragraph('second item in unordered list',style = 'List Bullet')
# 添加有序列表  数字
document.add_paragraph('first item in ordered list',style = 'List Number')
document.add_paragraph('second item in ordered list',style = 'List Number')
# 添加分节符，分节符是页面分割线，用于分隔不同内容，相当于翻一页
document.add_section()

records = (
('严明轩','男',19),
('卜士函','男',100)
)
# 添加图片，  路径  宽度
document.add_picture(r"C:\Users\周钟启\Pictures\Screenshots\1.png",width = Cm(5.2))


# 添加表格
table = document.add_table(rows = 1,cols = 3)
# 设置表格样式
table.style = 'Dark List'
# 设置表格标题
"""
方法/属性	                  作用
add_table(rows, cols)	    创建表格
table.rows[0]	            获取第一行
add_row()	                添加新行
.text	                    设置单元格文字
str()	                    将数字转为字符串"""
# 每个单元格 rows[index].cells[index]
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Name'
hdr_cells[1].text = 'Gender'
hdr_cells[2].text = 'Age'
# 添加表格内容  为表格添加行
for name,gender,age in records:
    # 添加行
    row_cells = table.add_row().cells
    # 设置表头
    row_cells[0].text = name
    row_cells[1].text = gender
    row_cells[2].text = str(age)

# 添加分页符
document.add_page_break()
# 保存文档
document.save('05-demo-Word.docx')
```

```python
"""
可以生成模板文件
"""
from docx import Document
from docx.document import Document as Doc

# 将真实信息用字典的方式保存在列表中
employees = [
    {
        'name': '骆昊',
        'id': '100200198011280001',
        'sdate': '2008年3月1日',
        'edate': '2012年2月29日',
        'department': '产品研发',
        'position': '架构师',
        'company': '成都华为技术有限公司'
    },
    {
        'name': '王大锤',
        'id': '510210199012125566',
        'sdate': '2019年1月1日',
        'edate': '2021年4月30日',
        'department': '产品研发',
        'position': 'Python开发工程师',
        'company': '成都谷道科技有限公司'
    },
    {
        'name': '李元芳',
        'id': '2102101995103221599',
        'sdate': '2020年5月10日',
        'edate': '2021年3月5日',
        'department': '产品研发',
        'position': 'Java开发工程师',
        'company': '同城企业管理集团有限公司'
    },
]
# 对列表进行循环遍历，批量生成Word文档
for emp_dict in employees:
    # 读取离职证明模板文件
    doc = Document('resources/离职证明模板.docx')  # type: Doc
    # 循环遍历所有段落寻找占位符
    for p in doc.paragraphs:
        if '{' not in p.text:
            continue
        # 不能直接修改段落内容，否则会丢失样式
        # 所以需要对段落中的元素进行遍历并进行查找替换
        for run in p.runs:
            if '{' not in run.text:
                continue
            # 将占位符换成实际内容
            # find 返回出现的标签，从零开始
            start, end = run.text.find('{'), run.text.find('}')
            # key 不包含{}  place_holder 包含{}
            key, place_holder = run.text[start + 1:end], run.text[start:end + 1]
            # 实现占位符的替换，emp_dict[key]获取对应的值
            run.text = run.text.replace(place_holder, emp_dict[key])
    # 每个人对应保存一个Word文档
    doc.save(f'{emp_dict["name"]}离职证明.docx')
```
