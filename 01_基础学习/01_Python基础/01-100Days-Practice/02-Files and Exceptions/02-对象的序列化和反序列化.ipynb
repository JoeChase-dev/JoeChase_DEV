```python
"""
JSON (JavaScript Object Notation)概述
在 Python 中，我们可以将程序中的数据以 JSON 格式进行保存。
原因：因为它结构紧凑而且是纯文本，任何操作系统和编程语言都能处理纯文本，这就是实现跨语言跨平台数据交换的前提条件

Python	                    JSON
dict	                object
list / tuple	        array
str	                    string
int / float 	        number
bool （True / False）	boolean (true / false)
None	                    null
"""
```

```python
"""读取JSON格式的数据"""
import json
my_dict = {
    'name':'罗浩',
    'age': 20,
    'friends':['小明', '小红'],
    'cars':[
        {'brand':'宝马', 'max_speed':240},
        {'brand':'奔驰', 'max_speed':220},
        {'brand':'奥迪', 'max_speed':280}
    ]
}
print(json.dumps(my_dict))
```

```python
"""将字典处理成JSON格式，并写入文本文件"""
# 将dumps函数换为dump函数，并且传入文件对象即可
"""
四个重要的函数：
dump    :将Python对象转换为JSON格式，并写入文件
dumps   :将Python对象转换为JSON格式，并返回JSON字符串
load    :将JSON格式字符串转换为Python对象，并读取文件
loads   :将JSON格式字符串转换为Python对象，并返回Python对象

序列化与反序列化：
序列化：将对象转换为字节流的过程
反序列化：将字节流转换为对象的过程
维基百科上的解释是：“序列化（serialization）在计算机科学的数据处理中，是指将数据结构或对象状态转换为可以存储或传输的形式，这样在需要的时候能够恢复到原先的状态，而且通过序列化的数据重新获取字节时，可以利用这些字节来产生原始对象的副本（拷贝）。与这个过程相反的动作，即从一系列字节中提取数据结构的操作，就是反序列化（deserialization）”。

"""
import json
my_dict = {
    'name':'罗浩',
    'age': 20,
    'friends':['小明', '小红'],
    'cars':[
        {'brand':'宝马', 'max_speed':240},
        {'brand':'奔驰', 'max_speed':220},
        {'brand':'奥迪', 'max_speed':280}
    ]
}
with open('data.json', 'w') as file:
    json.dump(my_dict,file)


with open('data.json', 'r') as file:
    data = json.load(file)
print(data)

with open('data.json', 'r') as file:
    my_data = json.load(file)
    print(type(my_data))
    print(my_data)
```
