```python
# 排序算法（选择、冒泡和归并）和查找算法（顺序和折半）
def select_sort(items, comp=lambda x, y: x < y):
    """简单选择排序"""
    # items = items[:] 是 Python 中的列表切片复制写法。
    # 等价于 items.copy() 或 list(items)
    items = items[:]
    # 选择排序，每次选择一个最小的元素，与第一个元素交换位置

    for i in range(len(items) - 1):
        min_index = i
        for j in range(i + 1, len(items)):
            if comp(items[j], items[min_index]):
                min_index = j
        items[i], items[min_index] = items[min_index], items[i]
    return items


def bubble_sort(items, comp=lambda x, y: x > y):
    """冒泡排序"""
    items = items[:]
    for i in range(len(items) - 1):
        swapped = False
        for j in range(len(items) - 1 - i):
            if comp(items[j], items[j + 1]):
                items[j], items[j + 1] = items[j + 1], items[j]
                swapped = True
        if not swapped:
            break
    return items


def bubble_sort(items, comp=lambda x, y: x > y):
    """搅拌排序(冒泡排序升级版)"""
    items = items[:]
    for i in range(len(items) - 1):
        swapped = False
        for j in range(len(items) - 1 - i):
            if comp(items[j], items[j + 1]):
                items[j], items[j + 1] = items[j + 1], items[j]
                swapped = True
        if swapped:
            swapped = False
            for j in range(len(items) - 2 - i, i, -1):
                if comp(items[j - 1], items[j]):
                    items[j], items[j - 1] = items[j - 1], items[j]
                    swapped = True
        if not swapped:
            break
    return items


def merge(items1, items2, comp=lambda x, y: x < y):
    """合并(将两个有序的列表合并成一个有序的列表)"""
    items = []
    index1, index2 = 0, 0
    while index1 < len(items1) and index2 < len(items2):
        if comp(items1[index1], items2[index2]):
            items.append(items1[index1])
            index1 += 1
        else:
            items.append(items2[index2])
            index2 += 1

    # 切片 items1[index1:] — 切尾巴
    # 表示：从 index1 位置开始，一直取到列表末尾
    # items1[index1:len(items1)]   # 完整写法
    # items1[index1:]              # 省略 end，默认到末尾

    items += items1[index1:]
    items += items2[index2:]
    return items

# 专业级代码组织方式
# 入口函数
def merge_sort(items, comp=lambda x, y: x < y):
    #list(items) 是 Python 中的列表切片复制写法。
    # 等价于 items.copy() 或 list(items)
    # 设计上保持纯函数风格——不修改输入，只返回新结果。
    return _merge_sort(list(items), comp)

#内部递归函数
# def merge_sort(...):      # 公开接口 → 用户直接调用的
# def _merge_sort(...):     # 私有实现 → 内部使用的
"""
前缀	                含义	                        类比
无前缀	        public —— 公共 API	            餐厅的门面
_ 单下划线	    protected/internal —— 内部使用	厨房内部
__ 双下划线	    private —— 强制名称改写	        后厨密室
"""
def _merge_sort(items, comp):
    """归并排序"""
    if len(items) < 2:
        return items
    mid = len(items) // 2
    # 先通过 _merge_sort() 拆成一个个单元素列表,然后依次通过 merge() 两两进行结合成一个有序列表
    left = _merge_sort(items[:mid], comp)
    right = _merge_sort(items[mid:], comp)
    return merge(left, right, comp)
# 这种设计模式叫做 Facade Pattern（门面模式） 或 Wrapper Function（包装函数）。


def seq_search(items, key):
    """顺序查找"""
    # enumerate 更 Pythonic、更优雅！
    for index, item in enumerate(items):
        if item == key:
            return index
    return -1


def bin_search(items, key):
    # 条件为有序序列
    """折半查找"""
    start, end = 0, len(items) - 1
    # 若 key 不存在 则start > end
    while start <= end:
        mid = (start + end) // 2
        if key > items[mid]:
            start = mid + 1
        elif key < items[mid]:
            end = mid - 1
        else:
            return mid
    return -1
```

```python
"""常用算法：

穷举法 - 又称为暴力破解法，对所有的可能性进行验证，直到找到正确答案。
贪婪法 - 在对问题求解时，总是做出在当前看来最好的选择，不追求最优解，快速找到满意解。
分治法 - 把一个复杂的问题分成两个或更多的相同或相似的子问题，再把子问题分成更小
        的子问题，直到可以直接求解的程度，最后将子问题的解进行合并得到原问题的解
回溯法 - 回溯法又称为试探法，按选优条件向前搜索，当搜索到某一步发现原先选择并不
        优或达不到目标时，就退回一步重新选择。
动态规划 - 基本思想也是将待求解问题分解成若干个子问题，先求解并保存这些子问题的
        解，避免产生大量的重复运算。"""
```

```python
"""贪婪法
"""
"""
贪婪法：在对问题求解时，总是做出在当前看来是最好的选择，不追求最优解，快速找到满意解。
输入：
20 6
电脑      200     20
收音机     20      4
钟       175     10
花瓶      50      2
书       10      1
油画      90      9
"""
class Thing(object):
    """物品"""
    # 构造函数，在创建对象时，自动调用
    def __init__(self, name, price, weight):
        self.name = name
        self.price = price
        self.weight = weight
#   @property  # 将方法变成属性调用
    @property
    def value(self):
        """价格重量比"""
        return self.price / self.weight


def input_thing():
    """输入物品信息"""
    # input得到字符串，split进行分割，得到列表，然后进行解包
    name_str, price_str, weight_str = input().split()
    # 返回值为元组
    return name_str, int(price_str), int(weight_str)


def main():
    """主函数"""

    # map(函数, 可迭代对象)
    # 作用：将可迭代对象的每一个元素依次传入函数，返回结果的迭代器
    # 例：map(lambda x: x * x, [1, 2, 3, 4, 5, 6, 7, 8, 9])
    # 输出：[1, 4, 9, 16, 25, 36, 49, 64, 81]
    # 等价于列表推导式：  [int(x) for x in input().split()]
    max_weight, num_of_things = map(int, input().split())
    all_things = []
    for _ in range(num_of_things):
#         第1步: input_thing() 返回元组
#        ('电脑', 200, 20)
#
#       第2步: * 操作符将元组 "解包" 为独立的参数
#        *('电脑', 200, 20)
#
#        等价于:
#        Thing('电脑', 200, 20)
#
#       第3步: 调用 Thing 的 __init__
#        → 创建一个 Thing 对象
        all_things.append(Thing(*input_thing()))

        # list.sort(key=排序依据函数, reverse=True/False)
        # key	用什么来比较	lambda x: x.value
        # reverse	是否降序	True=从大到小
    all_things.sort(key=lambda x: x.value, reverse=True)
    total_weight = 0
    total_price = 0
    for thing in all_things:
        if total_weight + thing.weight <= max_weight:
            print(f'小偷拿走了{thing.name}')
            total_weight += thing.weight
            total_price += thing.price
    print(f'总价值: {total_price}美元')


# if __name__ == '__main__':                     ║
# ║       main()                                    ║
# ║                                                  ║
# ║   ─────────────────────────────────────          ║
# ║   本质: 一个"开关"                               ║
# ║                                                  ║
# ║   直接运行 → 开关 ON  → 执行 main()             ║
# ║   被导入   → 开关 OFF → 跳过,只提供定义          ║
# ║                                                  ║
# ║   目的:
# ║   避免"导入时就触发副作用"
# ║   ✓ 让同一个文件既能运行也能被导入               ║
# ║   ✓ 防止"导入时就触发副作用"(如弹窗/等待输入等)  ║
# ║   ✓ 是 Python 社区的标准惯例(PEP推荐)           ║
# ║                                                  ║
# ║   记忆口诀:                                     ║
# ║   "如果我是主角(__main__),就开始表演(main())"
if __name__ == '__main__':
    main()
"""
所有语法点汇总表
语法	                        所在行	         含义
class Thing(object)	        333	             定义类，继承自 object
def __init__(self, ...)	    336	             构造方法，初始化对象属性
self.name = name	        337	             绑定实例属性
@property	                341	            装饰器，让方法变属性
*input_thing()	            365	            解包操作符，元组→多参数
lambda x: x.value	        366	            匿名函数，定义排序键
.sort(key=..., reverse=True)366	            列表原地排序
f-string f'{var}'	        371,374	        格式化字符串
if __name__ == '__main__':	377	            程序入口保护
"""
```

```python
# 分治法例子
"""
快速排序 - 选择枢轴对元素进行划分，左边都比枢轴小右边都比枢轴大
"""
def quick_sort(items, comp=lambda x, y: x <= y):
    items = list(items)[:]
    _quick_sort(items, 0, len(items) - 1, comp)
    return items


def _quick_sort(items, start, end, comp):
    if start < end:
        pos = _partition(items, start, end, comp)
        _quick_sort(items, start, pos - 1, comp)
        _quick_sort(items, pos + 1, end, comp)


def _partition(items, start, end, comp):
    pivot = items[end]
    i = start - 1
    for j in range(start, end):
        if comp(items[j], pivot):
            i += 1
            items[i], items[j] = items[j], items[i]
    items[i + 1], items[end] = items[end], items[i + 1]
    return i + 1
```

```python
# 回溯法例子：骑士巡逻。

"""
递归回溯法：叫称为试探法，按选优条件向前搜索，当搜索到某一步，发现原先选择并不优或达不到目标时，就退回一步重新选择，比较经典的问题包括骑士巡逻、八皇后和迷宫寻路等。
"""
import sys
import time

SIZE = 5
total = 0

# 打印棋盘
def print_board(board):
    for row in board:
        for col in row:
            print(str(col).center(4), end='')
        print()


def patrol(board, row, col, step=1):
    if row >= 0 and row < SIZE and \
        col >= 0 and col < SIZE and \
        board[row][col] == 0:
        board[row][col] = step
        if step == SIZE * SIZE:
            global total
            total += 1
            print(f'第{total}种走法: ')
            print_board(board)
        patrol(board, row - 2, col - 1, step + 1)
        patrol(board, row - 1, col - 2, step + 1)
        patrol(board, row + 1, col - 2, step + 1)
        patrol(board, row + 2, col - 1, step + 1)
        patrol(board, row + 2, col + 1, step + 1)
        patrol(board, row + 1, col + 2, step + 1)
        patrol(board, row - 1, col + 2, step + 1)
        patrol(board, row - 2, col + 1, step + 1)
        board[row][col] = 0


def main():
    board = [[0] * SIZE for _ in range(SIZE)]
    patrol(board, SIZE - 1, SIZE - 1)


if __name__ == '__main__':
    main()
```

```python
"""动态规划例子：子列表元素之和的最大值。

说明：子列表指的是列表中索引（下标）连续的元素构成的列表；列表中的元素是int类型，可能包含正整数、0、负整数；程序输入列表中的元素，输出子列表元素求和的最大值，例如：

输入：1 -2 3 5 -3 2

输出：8

输入：0 -2 3 5 -1 2

输出：9

输入：-9 -2 -3 -5 -3

输出：-2"""

def main():
    items = list(map(int, input().split()))
    overall = partial = items[0]
    for i in range(1, len(items)):
        partial = max(items[i], partial + items[i])
        overall = max(partial, overall)
    print(overall)


if __name__ == '__main__':
    main()
# 说明：这个题目最容易想到的解法是使用二重循环，但是代码的时间性能将会变得非常的糟糕。使用动态规划的思想，仅仅是多用了两
# 个变量，就将原来$O(N^2)$复杂度的问题变成了$O(N)$。
```

```python
"""
装饰器函数
使用和取消装饰器"""

# 功能：这是一个计时装饰器，可以给任意函数"装上"计时功能，自动打印该函数的执行耗时
from time import time
# 外层装饰器工厂,实现函数调用的时间记录
def record_time(func):
    """装饰器函数，记录函数执行时间"""
    # 内层
    # *args — 收集所有位置参数 (1, 2, 3)
    # **kwargs — 收集所有关键字参数 {name="张三", age=20}
    def wrapper(*args, **kwargs):
        start = time()
        # 记录原函数的返回值结果
        result = func(*args, **kwargs)
        # 计算并打印函数执行时间
        print(f'{func.__name__}:{time() - start}秒')
        # 返回原函数的返回值结果
        return result
    return wrapper
```

```python
# 如果装饰器不希望跟print函数耦合，可以编写可以参数化的装饰器。
# 参数化装饰器	把输出方式当参数传入，灵活多变
# 耦合 = 绑死了、分不开
# 解耦	让各部分独立，方便替换
from functools import wraps
from time import time


def record(output): # # output 是一个参数，可以是任何函数
    """可以参数化的装饰器"""

    def decorate(func):

        @wraps(func)
        def wrapper(*args, **kwargs):
            start = time()
            result = func(*args, **kwargs)
            output(func.__name__, time() - start)
            return result

            return wrapper

	return decorate



```

```python
"""
案例：用装饰器实现单例模式
"""
"""
概念	                                                        说明
闭包 (Closure)	            wrapper 函数引用了外层函数的 instances 字典。即使 singleton() 执行完毕，instances 也
                           会被销毁，因为 wrapper 仍持有对它的引用
instances = {}	            这是一个字典，以类本身 为键，以该类的唯一实例为值。每个被 @singleton 装饰的类都会在各自对
                            的闭包中有独立的 instances 字典
if cls not in instances	     判断当前类是否已经有过实例化记录"""
# wraps 是一个工具装饰器，用于保留被装饰函数/类的元信息（如 __name__, __doc__, __module__ 等）。如果不使用它，被装饰后的
# 类会丢失原始的类名和文档字符串。
from functools import wraps
def singleton(cls):
    """单例装饰器"""
    # ① 类级别的缓存字典
    instances = {}
    # ② 保留原类的元信息
    @wraps(cls)
    def wrapper(*args,**kwargs):
        # ③ 如果类不在缓存字典中，则创建新的实例，并添加到缓存字典中
        if cls not in instances:
            instances[cls] = cls(*args,**kwargs)
        # ⑥ 后续调用：直接返回缓存的实例
        return instances[cls]

    return wrapper

@singleton
class President:
    """总统类"""
    pass


# 等价于： President = singleton(President)
"""
 这种方式的优缺点
        优点	                        缺点
✅ 代码简洁优雅	            ❌ 多线程环境下不安全（可能创建多个实例）
✅ 复用性强，可装饰任意类	       ❌ 子类继承时需要额外处理
✅ 对原有类代码零侵入	        ❌ 无法通过 __new__ 或 __init__ 再次初始化
✅ 利用闭包实现状态保持"""
```

```python
"""
线程安全的单例装饰器

RLock（Reentrant Lock，可重入锁）与普通 Lock 的关键区别：

特性          	Lock	                RLock
可重入	❌ 同一线程再次获取会死锁	        ✅ 同一线程可多次 acquire
计数器	无	                            有，acquire+1，release-1
适用场景	简简单的互斥	                    可能发生递归/嵌套调用的场景

选择 RLock 而非 Lock 是为了防止在极端情况下（如 __init__ 中又触发了某些逻辑导致锁竞争）产生死锁。
"""
from functools import wraps
from threading import RLock


def singleton(cls):
    """线程安全的单例装饰器"""
    instances = {}
    # 线程锁（也是闭包自由变量)
    locker = RLock()

    @wraps(cls)
    def wrapper(*args, **kwargs):
        # 第一重检查（外层）
        if cls not in instances:
            #加锁
            with locker:
                # 第二重检查（内层）
                if cls not in instances:
                    instances[cls] = cls(*args, **kwargs)
        return instances[cls]

    return wrapper
```

```python
"""
Python 锁
"""
# 1. 锁的语法
# Python 锁在 threading 模块中实现，常用两种：
# threading.Lock（互斥锁）
from threading import Lock
lock = Lock()

# 方式1：手动 acquire/release
lock.acquire()
try:
    # 临界区代码
    ...
finally:
    lock.release()

# 方式2：推荐用 with 上下文管理（自动释放）
with lock:
    # 临界区代码
    ...

# threading.RLock（可重入锁）
from threading import RLock
rlock = RLock()

with rlock:
    # 同一线程可再次进入
    with rlock:
        ...
# Lock 内部维护持有线程标识和计数，只有计数归零才真正释放。

# 所的作用
"""核心作用：在多线程环境下，保证同一时刻只有一个线程执行某段代码（临界区），防止竞态条件（Race Condition）导致数据错乱。"""
# 竞态条件原理
counter = 0
def inc():
    global counter
    for _ in range(100000):
        counter += 1   # 非原子操作：读→改→写
# 多线程并发时，两个线程可能同时读相同值，各自+1写回，最终结果小于预期。
# 加锁后：
# 保证每次只有一个线程做 +=1
lock = Lock()
def inc_safe():
    global counter
    for _ in range(100000):
        #  会让一个线程完整做完 counter += 1 的 读→改→写 三步后，另一个线程才能开始。
        with lock:
            counter += 1

"""
3. 应用场景
场景	                说明	                            推荐锁类型
共享变量更新	        多线程读写全局变量、列表、字典等	    Lock/RLock
单例模式（线程安全）	确保类只有一个实例	                RLock（或Lock）+双重检查
文件/数据库写入	    防止并发写冲突	                Lock
递归函数内加锁	    同一线程可能多次进入临界区	        RLock
生产者-消费者队列	    保证队列操作完整性	Lock 配合        Condition
资源池（连接池、线程池）限制同时访问资源的线程数	        Lock/Semaphore
"""

"""
5. 锁的选择原则
无重入需求 → 用 Lock（更轻量）
有递归/嵌套加锁 → 用 RLock
高并发读、少量写 → 考虑 threading.RLock + 读写分离策略（如 reader-writer lock）或 queue.Queue 等高级结构
"""
"""
原子操作（Atomic Operation）是指不可被中断的操作，它在执行过程中不会被其他线程或进程打断，要么全部完成，要么完全没执行，不会出现“做到一半被切换”的情况。
"""
```

```python
"""
面向对象相关知识
"""
# 三大支柱：封装、继承、多态

 # 例子：工资结算系统。

"""
月薪结算系统 - 部门经理每月15000 程序员每小时200 销售员1800底薪加销售额5%提成
"""
#  Python 抽象基类（Abstract Base Class, ABC）
from abc import ABCMeta, abstractmethod
# ABCMeta 元类，声明该类是"抽象类"，不能直接实例化
# abstractmethod 装饰器，标记"抽象方法"，子类必须实现

#metaclass=ABCMeta 元类声明，表示 Employee 是一个抽象类，不能直接实例化。
class Employee(metaclass=ABCMeta):
    """员工(抽象类)"""

    def __init__(self, name):
        self.name = name
    # 强制子类必须实现
    @abstractmethod
    def get_salary(self):
        """结算月薪(抽象方法)"""
        pass


class Manager(Employee):
    """部门经理"""

    def get_salary(self):
        return 15000.0


class Programmer(Employee):
    """程序员"""

    def __init__(self, name, working_hour=0):
        self.working_hour = working_hour
        super().__init__(name)

    def get_salary(self):
        return 200.0 * self.working_hour


class Salesman(Employee):
    """销售员"""

    def __init__(self, name, sales=0.0):
        self.sales = sales
        super().__init__(name)

    def get_salary(self):
        return 1800.0 + self.sales * 0.05


class EmployeeFactory:
    """创建员工的工厂（工厂模式 - 通过工厂实现对象使用者和对象之间的解耦合）"""

    @staticmethod
    def create(emp_type, *args, **kwargs):
        """创建员工"""
        all_emp_types = {'M': Manager, 'P': Programmer, 'S': Salesman}
        cls = all_emp_types[emp_type.upper()]
        return cls(*args, **kwargs) if cls else None


def main():
    """主函数"""
    emps = [
        EmployeeFactory.create('M', '曹操'),
        EmployeeFactory.create('P', '荀彧', 120),
        EmployeeFactory.create('P', '郭嘉', 85),
        EmployeeFactory.create('S', '典韦', 123000),
    ]
    for emp in emps:
        print(f'{emp.name}: {emp.get_salary():.2f}元')


if __name__ == '__main__':
    main()
```

```python
"""
__repr__ — 定义对象的官方字符串表示
    用于控制 print() / 交互式终端 / 调试时 对象显示成什么样子。

__repr__	print(obj)、交互终端直接输入对象、调试	            面向开发者，用于调试
__str__	    str(obj)、f'{obj}'、print(obj)（无 __repr__ 时）	面向用户，用于友好展示
"""
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __repr__(self):
        return f'Student({self.name}, {self.score})'      # 给开发者看

    def __str__(self):
        return f'{self.name}: {score}分'                   # 给用户看


s = Student("张三", 95)

print(s)           # → 调用 __str__ → "张三: 95分"
repr(s)            # → 调用 __repr__ → "Student(张三, 95)"
# 交互式终端输入 s  → 调用 __repr__
# Python 会优先使用 __str__；如果没定义 __str__，则回退用 __repr__
```

```python
"""
@property — 定义属性
核心作用：调用时不需要加括号 ()
"""
poker = Poker()

# ❌ 没有加 @property 时 — 必须当方法调用（带括号）
print(poker.has_more())          # True

# ✅ 加了 @property 后 — 当属性访问（无括号）
print(poker.has_more)            # True   ← 不用写 ()

# 更多功能：控制读写权限
class Student:
    def __init__(self, score):
        self._score = score       # 下划线约定：内部属性

    @property                     # getter — 读取
    def score(self):
        return self._score

    @score.setter                 # setter — 设置（带校验逻辑）
    def score(self, value):
        if not 0 <= value <= 100:
            raise ValueError("成绩必须在0~100之间！")
        self._score = value


s = Student(85)

print(s.score)           # 85     → 调用 getter

s.score = 95             # ✅ 正常 → 调用 setter
s.score = 150            # → ValueError! 成绩必须在0~100之间！
# 好处：
#   对外暴露"看起来像普通属性"的接口
#   内部可以加入验证、计算、日志等逻辑
#   未来需求变更时，可以在不改变外部调用方式的情况下修改实现
```

```python
# Python 对象复制与内存管理详解
# 这段内容涵盖了 对象拷贝 和 垃圾回收 两个核心话题，我逐一拆解：
"""
一、深拷贝 vs 浅拷贝（概念对比）
                浅拷贝 (Shallow Copy)	    深拷贝 (Deep Copy)
复制的是什么？	只复制容器本身	        递归复制所有嵌套对象
内部元素	        共享引用同一对象	        创建全新的独立副本
copy.copy()	    ✅	                    ❌
copy.deepcopy()	❌	                    ✅
"""
# 图解理解
import copy

# 原始数据
a = [1, 2, [3, 4]]

b = copy.copy(a)      # 浅拷贝
c = copy.deepcopy(a)  # 深拷贝

a[2].append(99)

print(b)  # [1, 2, [3, 4, 99]]  ← 受影响！因为 b[2] 和 a[2] 指向同一个子列表
print(c)  # [1, 2, [3, 4]]      ← 不受影响！完全独立的副本
"""
浅拷贝:                    深拷贝:
  a ──┐                     a ──→ [1, 2, ─┐]
      ↓                                   │
    [1,2,┐]                 c ──→ [1, 2, ─┤]
         ↓                                ↓
  b ──→ [1,2,┐]          各自独立的新列表 → [3, 4]
             ↓                    ↘
           [3,4] ← 共享同一个！     [3,4]  ← 完全独立的新列表"""
# 一句话总结：浅拷贝是"外壳新、里子旧"；深拷贝是"连皮带肉全新"。

# 二、垃圾回收机制 —— 引用计数
# Python 用 引用计数 作为主要内存管理方式：

"""c
typedef struct _object {
    int ob_refcnt;               // 引用计数器
    struct _typeobject *ob_type; // 类型信息
} PyObject;
每个对象内部都有一个计数器 ob_refcnt，记录有多少个变量在引用它。
"""

"""
引用计数 +1 的情况（4种）
场景	        示例	            解释
创建对象	    a = 23	    新对象诞生，计数=1
赋值引用	    b = a	    多了一个名字指向它
传入函数	    f(a)	    函数参数也是一次引用
存入容器	    list1 = [a, a]	列表内部也持有引用


引用计数    -1 的情况（5种）
场景	        示例	                解释
    del 显式删除	del a	    断开引用
    重新赋值	    a = 24	    原来的23少了一个引用
离开作用域	函数执行完毕	    局部变量自动销毁
容器被销毁	del list1	    容器内元素的引用都-1
从容器移除	list.remove(a)	    少了一个持有者
当引用计数降到 0 时，Python 立即回收该对象的内存。
"""

import sys

a = "hello"
print(sys.getrefcount(a))  # 输出 2 （a 自己 + getrefcount 的临时参数）

b = a                      # 再多一个引用
print(sys.getrefcount(a))  # 输出 3

del b                      # 减少一个
print(sys.getrefcount(a))  # 回到 2


# 三、循环引用问题 —— 引用计数的致命缺陷
# 这是引用计数最大的漏洞：


list1 = []
list2 = []

list1.append(list2)   # list1 引用了 list2
list2.append(list1)   # list2 又引用了 list1

del list1             # 删除外部引用
del list2             # 删除外部引用
# 此时的内存状态：

"""
list1 对象 ──────→ list2 对象
  ↑__________________↓
"""
"""
互相引用！
结果：两个对象的引用计数都是 1（互相引用），永远不会降为 0，导致 内存泄漏！

四、Python 的解决方案（两套辅助机制）
方案 1：标记 - 清除（Mark-Sweep）
定期扫描所有对象，标记哪些是"可达的"，然后清除不可达的对象。

从根对象（全局变量、栈上局部变量）出发
  ↓
能遍历到的 → 标记为"存活"
遍历不到的 → 即使引用计数 > 0，也视为垃圾 → 回收



方案 2：分代收集（Generational Collection）
基于一个经验规律：大多数对象生命周期都很短。

第0代（新生代）: 新建的对象放这里，频繁检查垃圾
    ↓ 存活后晋升
第1代（中年代）: 检查频率降低
    ↓ 继续存活
第2代（老年代）: 很少检查（大概率会一直活着）

"""
"""
五、弱引用（weakref）—— 手动避免循环引用
如果两个对象需要互相知道对方的存在，但又不想造成循环引用，就用 弱引用：
"""


import weakref

class Node:
    def __init__(self, value):
        self.value = value
        self.other = None  # 先设为 None

a = Node("A")
b = Node("B")

# 强引用会导致循环引用:
# a.other = b
# b.other = a  # 💀 内存泄漏风险

# 弱引用方案: 不增加引用计数！
a.other = weakref.ref(b)       # b 的弱引用
b.other = weakref.ref(a)       # a 的弱引用

# 使用时通过 .() 调用获取真实对象
print(a.other().value)         # 输出 "B"

del a, b                       # 两个对象都能正常回收了 ✓
"""code
强引用循环:                  弱引用循环:
  A ──强──→ B                 A ──强──→ B
  ↑_______↓                   A ←─弱── B  （弱引用不计入 refcnt）
  refcnt=1 永不归零           refcnt 都能归零 ✓
核心区别：弱引用只是"知道对方在哪"，但不"抓住不放"，不影响对方的引用计数。"""

"""总结脑图
code
对象复制
 ├── 浅拷贝: 复制外壳，内部共享
 └── 深拷贝: 完全独立的新副本

垃圾回收
 ├── 主力: 引用计数（+1/-1 规则）
 ├── 缺陷: 循环引用导致内存泄漏
 └── 补救:
      ├── 标记-清除: 定期扫描不可达对象
      ├── 分代收集: 分代管理，新生代频繁检查
      └── 弱引用: 开发者手动避免循环引用"""
```

```python
"""
一、什么是"魔法方法"？
以 __ 开头和结尾的方法，也叫 双下方法（dunder methods）。它们是 Python 的 钩子——你不需要直接调用，Python 在特定场景下自动调用它们。"""

# 二、常用魔法方法分类速查表
# 1️⃣ 表示相关（让对象"能被打印"
"""
        触发时机                    示例
_str__	print(obj) / str(obj)	面向用户的友好字符串
__repr__	repr(obj) / 交互式终端输出	面向开发者的精确描述
"""
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __str__(self):
        return f'【学生】{self.name}，成绩：{self.score}'     # 给用户看

    def __repr__(self):
        return f'Student(name={self.name!r}, score={self.score})'  # 给开发者调试用

s = Student("张三", 95)
print(s)          # → 【学生】张三，成绩：95      （调用 __str__）
print(repr(s))    # → Student(name='张三', score=95) （调用 __repr__）
# 如果没定义 __str__，print 会 fallback(撤退) 到 __repr__

# 2️⃣ 运算符重载（回答问题①：自定义对象能用运算符吗？）
"""
方法	                对应运算符	      说明
__add__	        a + b	            加法
__sub__	        a - b	            减法
__mul__	        a * b	            乘法
__truediv__	    a / b	            除法
__eq__	        a == b	            等于
__lt__, __le__	<, <=	            小于、小于等于
__gt__, __ge__	>, >=	            大于、大于等于
__len__	        len(obj)	        长度
__getitem__	    obj[key]	        索引访问
__setitem__	    obj[key] = val	    索引赋值
__contains__	x in obj	        成员判断
"""
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):           # 实现 + 运算
        return Vector(self.x + other.x, self.y + other.y)

    def __eq__(self, other):            # 实现 == 比较
        return self.x == other.x and self.y == other.y

    def __repr__(self):                 # 好看的打印格式
        return f'Vector({self.x}, {self.y})'

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)       # → Vector(4, 6)    ✅ 自定义对象可以用 +
print(v1 == v2)      # → False           ✅ 自定义对象可以比较

# 3️⃣ 哈希与集合/字典键（回答问题②③）
"""
方法	            作用	                    关联场景
__hash__	返回对象的哈希值（整数）	作为 dict 键 / 放入 set 时需要
__eq__	    判断两个对象是否相等	        配合 __hash__ 使用
"""
class Card:
    def __init__(self, suit, rank):
        self.suit = suit    # 花色 ♠♥♣♦
        self.rank = rank    # 点数 A,2,3...K

    def __repr__(self):
        return f'{self.suit}{self.rank}'

    def __hash__(self):
        # 返回一个稳定的整数哈希值
        return hash((self.suit, self.rank))

    def __eq__(self, other):
        return self.suit == other.suit and self.rank == other.rank


c1 = Card("♠", "A")
c2 = Card("♠", "A")
c3 = Card("♥", "K")

print(c1 == c2)            # → True   （内容相同）
print(hash(c1) == hash(c2)) # → True   （哈希值相同）

# 问题②：自定义对象能放入 set 吗？
my_set = {c1, c2, c3}
print(my_set)              # → {♠A, ♥K}   ✅ 能！且自动去重了！（c1 和 c2 被视为同一个）

# 问题③：自定义对象能作为 dict 键吗？
card_count = {c1: 4, c3: 1}
print(card_count[c2])      # → 4          ✅ 能！c2 和 c1 哈希+相等，所以能找到！

"""
哈希 (hash)
 ├── 本质：任意数据 → 固定数字（门牌号）
 ├── 目的：快速定位 O(1)，不用一个个遍历
 └── 要求：相等的对象必须有相同的哈希值

集合 (set) 去重原理
 ├── 加入元素 → 算 hash → 定位桶
 ├── 桶里已有元素 → 用 eq 逐一比较
 ├── hash相同 + eq相同 → 视为重复，丢弃
 └── 需要同时实现 __hash__ 和 __eq__

字典 (dict) 键原理
 └── 与 set 完全一致，只不过桶里多存了个 value

铁律
 └── 重写 __eq__就必须重写 __hash__
     且两者逻辑必须保持一致！
"""
# 关键规则 ⚠️
# 如果你重写了 __eq__，就必须重写 __hash__！
#
# 原因：
#   set / dict 内部流程：
#     1. 先用 __hash__() 快速定位到"桶"
#     2. 再用 __eq__() 精确比对是否真的相等
#
# 如果两个对象 __eq__ 相等但 __hash__ 不同：
#   → dict 会认为它们不是同一个 key → 数据错乱！
#
# 反之亦然：__hash__ 相同但 __eq__ 不等：
#   → 会被错误地判定为重复元素

# 4️⃣ 上下文管理器（回答问题④）
"""
方法	            作用	            触发时机
__enter__	进入上下文时执行	with 语句开始时
__exit__	退出上下文时执行	with 语句结束时（无论正常还是异常）
"""
class Timer:
    """计时器上下文管理器"""
    import time

    def __enter__(self):
        self.start = time.time()
        print("⏱ 计时开始...")
        return self              # 返回值赋给 as 后面的变量

    def __exit__(self, exc_type, exc_val, exc_tb):
        elapsed = time.time() - self.start
        print(f"⏱ 计时结束，耗时 {elapsed:.3f} 秒")
        # 返回 False 表示不吞掉异常，True 表示异常被处理了
        return False


# 问题④：自定义对象能用 with 语法吗？
with Timer() as t:
    sum(range(10_000_000))       # 做一些耗时操作
    # 输出: ⏱ 计时开始...
    #       ⏱ 计时结束，耗时 0.xxx 秒    ✅ 可以！


# 实际应用 —— 文件/锁的自动管理：
class MyFile:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        print(f"📂 打开文件 {self.filename}")
        self.f = open(self.filename, self.mode)
        return self.f             # as f 拿到的就是这个

    def __exit__(self, exc_type, exc_val, exc_tb):
        print(f"💾 关闭文件 {self.filename}")
        self.f.close()            # 保证一定会关闭！
        return False


with MyFile('test.txt', 'w') as f:
    f.write('hello')
    # 无论中间是否抛异常，__exit__ 都会执行 → 文件一定关闭

```

```python
"""
混入（Mixin）

例子：自定义字典限制只有在指定的key不存在时才能在字典中设置键值对。
"""

class SetOnceMappingMixin:
    """自定义混入类"""
    __slots__ = () # # ① 空元组，阻止该类创建 __dict__，节省内存

# 关键点：

# __slots__ = () — 声明此类没有实例属性，纯功能类
# super() — 不硬编码调用 dict.__setitem__，而是委托给 MRO 链的下一个类（这就是 Mixin 的精髓）
# 只覆写 __setitem__ — 单一职责，只管"限制重复设置"这一件事

    def __setitem__(self, key, value):
        if key in self:
            raise KeyError(str(key) + ' already set')
        return super().__setitem__(key, value) # ③ 调用下一个类的 __setitem__


class SetOnceDict(SetOnceMappingMixin, dict):
    """自定义字典"""
    pass


my_dict= SetOnceDict()
try:
    my_dict['username'] = 'jackfrued'
    my_dict['username'] = 'hellokitty'
except KeyError:
    pass
print(my_dict)
```

```python
"""元编程和元类

对象是通过类创建的，类是通过元类创建的，元类提供了创建类的元信息。所有的类都直接或间接的继承自object，所有的元类都直接或间接的继承自type。

例子：用元类实现单例模式。"""

import threading


class SingletonMeta(type):
    """自定义元类  继承自 type"""

    def __init__(cls, *args, **kwargs):
        """
        当类 President 被【定义/创建】时调用
        cls → 正在被创建的类（即 President 本身）
        """
        cls.__instance = None  #  ① 给类挂上一个类属性，存唯一实例
        cls.__lock = threading.RLock()
        super().__init__(*args, **kwargs)

    def __call__(cls, *args, **kwargs):
        """
        当你写 President() 时触发
        相当于拦截了实例化过程
        """
        if cls.__instance is None:
            with cls.__lock:
                if cls.__instance is None:
                    cls.__instance = super().__call__(*args, **kwargs)
        return cls.__instance


# metaclass=SingletonMeta 的含义：
# 不要用默认的 type 来创建我，用 SingletonMeta 来创建我
class President(metaclass=SingletonMeta):
    """总统(单例类)"""
    """指定元类为 SingletonMeta"""
    pass

"""
总结对比
概念	                触发时机	        作用
cls.__new__	    实例化时	        创建实例对象（分配内存）
cls.__init__	__new__ 后	    初始化实例属性
元类 __call__	ClassName() 时	拦截整个实例化过程
元类 __init__	class 定义时	    干预类的创建

元类的本质：它是一个"类的模板"，让你能在类被创建时和类被实例化时插入自定义逻辑。单例、ORM（如 Django Model）、自动注册等高级框架功能都依赖它。"""
```

```python

"""面向对象设计原则

单一职责原则 （SRP）- 一个类只做该做的事情（类的设计要高内聚）
开闭原则 （OCP）- 软件实体应该对扩展开发对修改关闭
依赖倒转原则（DIP）- 面向抽象编程（在弱类型语言中已经被弱化）
里氏替换原则（LSP） - 任何时候可以用子类对象替换掉父类对象
接口隔离原则（ISP）- 接口要小而专不要大而全（Python中没有接口的概念）
合成聚合复用原则（CARP） - 优先使用强关联关系而不是继承关系复用代码
最少知识原则（迪米特法则，LoD）- 不要给没有必然联系的对象发消息
说明：上面加粗的字母放在一起称为面向对象的SOLID原则。"""

"""GoF设计模式

创建型模式：单例、工厂、建造者、原型
结构型模式：适配器、门面（外观）、代理
行为型模式：迭代器、观察者、状态、策略
例子：可插拔的哈希算法（策略模式）。"""

class StreamHasher:
    """哈希摘要生成器"""

    def __init__(self, alg='md5', size=4096):
        self.size = size
        alg = alg.lower()
        self.hasher = getattr(__import__('hashlib'), alg.lower())()

    def __call__(self, stream):
        return self.to_digest(stream)

    def to_digest(self, stream):
        """生成十六进制形式的摘要"""
        for buf in iter(lambda: stream.read(self.size), b''):
            self.hasher.update(buf)
        return self.hasher.hexdigest()

def main():
    """主函数"""
    hasher1 = StreamHasher()
    with open('Python-3.7.6.tgz', 'rb') as stream:
        print(hasher1.to_digest(stream))
    hasher2 = StreamHasher('sha1')
    with open('Python-3.7.6.tgz', 'rb') as stream:
        print(hasher2(stream))


if __name__ == '__main__':
    main()
```

```python
"""迭代器和生成器
迭代器是实现了迭代器协议的对象。

Python中没有像protocol或interface这样的定义协议的关键字。
"定义协议" = 给对象定规矩：你必须有哪些方法，才能算"这类东西"。
！！！！
        定义协议" = 提前写一份"能力清单"，规定一个对象必须拥有哪些方法/属性才能被归为某类事物。

        Java/Go 用 interface / protocol 关键字来写这份清单
        Python 没有这个关键字，但用 ABC 或 鸭子类型 来达到同样目的
！！！！

Python中用魔术方法表示协议。
__iter__和__next__魔术方法就是迭代器协议。"""
class Fib(object):
    """迭代器"""

    def __init__(self, num):
        self.num = num
        self.a, self.b = 0, 1
        self.idx = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.idx < self.num:
            self.a, self.b = self.b, self.a + self.b
            self.idx += 1
            return self.a
        raise StopIteration()

"""

你写：                    for value in Fib(3):
                              print(value)
                               │
          ┌──────────────────┘
          ▼
   ┌──────────────┐
   │  __iter__()   │ ◀── 第一步（只调用一次）
   │  return self  │     返回迭代器对象
   └──────┬───────┘
          │
          ▼
   ┌──────────────┐     ┌──────────────┐
   │  __next__()   │ ──▶│  返回 0       │ ──▶ print(0)
   └──────────────┘     └──────────────┘
          │
          ▼
   ┌──────────────┐     ┌──────────────┐
   │  __next__()   │ ──▶│  返回 1       │ ──▶ print(1)
   └──────────────┘     └──────────────┘
          │
          ▼
   ┌──────────────┐     ┌──────────────┐
   │  __next__()   │ ──▶│  返回 1       │ ──▶ print(1)
   └──────────────┘     └──────────────┘
          │
          ▼
   ┌──────────────┐     ┌──────────────┐
   │  __next__()   │ ──▶│  StopIter!    │ ──▶ 循环结束 🔚
   └──────────────┘     └──────────────┘
"""
# 生成器是语法简化版的迭代器。

def fib(num):
    """生成器"""
    a, b = 0, 1
    for _ in range(num):
        a, b = b, a + b
        yield a

"""生成器进化为协程。

生成器对象可以使用send()方法发送数据，发送的数据会成为生成器函数中通过yield表达式获得的值。这样，生成器就可以作为协程使用，协程简单的说就是可以相互协作的子程序。"""

def calc_avg():
    """流式计算平均值"""
    total, counter = 0, 0
    avg_value = None
    while True:
        value = yield avg_value
        total, counter = total + value, counter + 1
        avg_value = total / counter


gen = calc_avg()
next(gen)
print(gen.send(10))
print(gen.send(20))
print(gen.send(30))
```

```python
"""
Python 没有内置的 protocol / interface 关键字，但有自己的一套方式来实现同样的效果。
"""
# Python 的两种替代方案
# 方案一：鸭子类型（Duck Typing）— Python 最推荐的方式
# "如果它走起来像鸭子，叫起来像鸭子，那它就是鸭子。"
#
# 不声明、不检查、直接用：
class Duck:
    def fly(self):
        print("鸭子飞")

class Airplane:
    def fly(self):          # 不需要继承任何接口
        print("飞机飞")

class Whale:
    def swim(self):         # 没有 fly 方法
        print("鲸鱼游")

def make_it_fly(thing):
    thing.fly()             # 直接调用，不检查类型

make_it_fly(Duck())      # ✅ 鸭子飞
make_it_fly(Airplane())  # ✅ 飞机飞
make_it_fly(Whale())     # ❌ AttributeError! 运行时才报错
# 核心思想：不需要事先约定接口，只要对象有 fly() 方法就能用。


# 方案二：抽象基类 ABC（Abstract Base Class）— 需要正式约束时
from abc import ABC, abstractmethod

class Flyable(ABC):           # ← 用 ABC 代替 interface
    @abstractmethod            # ← 装饰器标记"子类必须实现"
    def fly(self):
        pass

class Duck(Flyable):          # ← 继承 ABC
    def fly(self):
        print("鸭子飞")

class Bird(Flyable):
    pass                       # ← 忘记实现 fly()

d = Duck()                     # ✅ 正常
b = Bird()                     # ❌ TypeError: Can't instantiate abstract class Bird
                               #   with abstract method fly

"""
        鸭子类型	            ABC
灵活性	⭐⭐⭐ 极高	    一般
类型约束	运行时报错	    实例化时就报错
适用场景	快速开发、脚本	大型项目、框架设计
Python哲学	更 Pythonic	给习惯静态语言的人用
"""

# 方案三：Protocol 类型提示（Python 3.8+）
# 接近protocol的东西：
from typing import Protocol

class Flyable(Protocol):       # ← typing.Protocol！
    def fly(self) -> None: ...

def make_it_fly(thing: Flyable):   # 类型检查器会验证有没有 fly()
    thing.fly()

class PaperPlane:
    def fly(self): print("纸飞机")

make_it_fly(PaperPlane())   # mypy 检查通过！（结构子类型）
                             # 无需继承 Flyable！


"""
┌──────────────────────────────────────────────────────┐
│              如何在 Python 中定义"接口"？                │
├──────────────────┬───────────────────┬────────────────┤
│                  │                   │                │
│   鸭子类型        │   ABC 抽象基类      │  typing.Protocol│
│                  │                   │                 │
│   不声明任何东西   │   class X(ABC)    │  class X(Protocol)
│   有方法就直接用   │   @abstractmethod │  (纯类型提示)     │
│                  │                   │                 │
│   运行时可能报错   │   实例化时就报错    │  mypy 静态检查报错│
│                  │                   │                 │
│   推荐日常使用     │   推荐 API 设计    │  推荐大型项目     │
└──────────────────┴───────────────────┴────────────────┘

注意：Python 没有 interface / protocol 【关键字】，
但有以上三种【机制】来达到同样目的。
"""
```

```python
"""
__iter__ 和 __next__  ----迭代器协议的核心方法。
"""
# 这两个方法必须配对实现
class Fib:
    """斐波那契数列迭代器"""

    def __init__(self, num):
        self.num = num           # 生成多少个数
        self.a = 0               # 当前值
        self.b = 1               # 下一个值
        self.idx = 0             # 当前计数

    def __iter__(self):          # ← 你选中的方法1
        return self

    def __next__(self):
        if self.idx < self.num:      # 还有元素吗？
            val = self.a             # 取当前值
            self.a, self.b = self.b, self.a + self.b  # 计算下一个
            self.idx += 1
            return val               # 返回值
        raise StopIteration          # 没有了！通知 for 循环停止



for i in Fib(5):
    print(i, end=" ")    # 输出: 0 1 1 2 3

# 二、__iter__ — "我是迭代器"     作用: 返回一个迭代器对象本身
# __iter__() 就是第 ① 步——告诉 Python"谁负责逐个产出元素"。

# 三、__next__ — "给你下一个值"   作用:每次被调用时返回下一个元素，没有更多元素时抛出 StopIteration


"""
五、两个角色的区别
Python 中有两个容易混淆的概念：

        可迭代对象 (Iterable)	            迭代器 (Iterator)
有	    __iter__()	                    __iter__() + __next__()
职责	    "可以被遍历"	                    "负责具体怎么遍历"
类比  	一本书	                        书签（记住读到哪了）
例子	    list, dict, str	                iter(list) 的结果
能重复遍历？✅ 可以多次从头来	        ❌ 只能用一次
"""
lst = [1, 2, 3]        # 列表是 Iterable（可迭代对象）
it = iter(lst)         # it 是 Iterator（迭代器）

list(it)   # [1, 2, 3]  ✅ 正常
list(it)   # []         ❌ 空的！迭代器已经耗尽


"""
六、一句话总结
方法	                一句话
__iter__	"把你自己交出来给 for 循环用" — 只调一次，返回迭代器自身
__next__	"给我下一个值" — 反复调用，直到抛出 StopIteration 为止
"""
"""
for 循环的本质就是不断调用 __next__() 取值的过程，而 __iter__() 就是这个过程的入场券。
"""
```

```python
"""并发编程
Python中实现并发编程的三种方案：多线程、多进程和异步I/O。并发编程的好处在于可以提升程序的执行效率以及改善用户体验；坏处在于并发的程序不容易开发和调试，同时对其他程序来说它并不友好。

多线程：Python中提供了Thread类并辅以Lock、Condition、Event、Semaphore和Barrier。Python中有GIL来防止多个线程同时执行本地字节码，这个锁对于CPython是必须的，因为CPython的内存管理并不是线程安全的，因为GIL的存在多线程并不能发挥CPU的多核特性。"""

"""
面试题：进程和线程的区别和联系？
进程 - 操作系统分配内存的基本单位 - 一个进程可以包含一个或多个线程
线程 - 操作系统分配CPU的基本单位
并发编程（concurrent programming）
1. 提升执行性能 - 让程序中没有因果关系的部分可以并发的执行
2. 改善用户体验 - 让耗时间的操作不会造成程序的假死
"""
import glob
import os
import threading

from PIL import Image

PREFIX = 'thumbnails'


def generate_thumbnail(infile, size, format='PNG'):
    """生成指定图片文件的缩略图"""
	file, ext = os.path.splitext(infile)
	file = file[file.rfind('/') + 1:]
	outfile = f'{PREFIX}/{file}_{size[0]}_{size[1]}.{ext}'
	img = Image.open(infile)
	img.thumbnail(size, Image.ANTIALIAS)
	img.save(outfile, format)


def main():
    """主函数"""
	if not os.path.exists(PREFIX):
		os.mkdir(PREFIX)
	for infile in glob.glob('images/*.png'):
		for size in (32, 64, 128):
            # 创建并启动线程
			threading.Thread(
				target=generate_thumbnail,
				args=(infile, (size, size))
			).start()


if __name__ == '__main__':
	main()


```

```python
# 多个线程竞争资源的情况。

"""
多线程程序如果没有竞争资源处理起来通常也比较简单
当多个线程竞争临界资源的时候如果缺乏必要的保护措施就会导致数据错乱
说明：临界资源就是被多个线程竞争的资源
"""
import time
import threading

from concurrent.futures import ThreadPoolExecutor


class Account(object):
    """银行账户"""

    def __init__(self):
        self.balance = 0.0
        self.lock = threading.Lock()

    def deposit(self, money):
        # 通过锁保护临界资源
        with self.lock:
            new_balance = self.balance + money
            time.sleep(0.001)
            self.balance = new_balance


def main():
    """主函数"""
    account = Account()
    # 创建线程池
    pool = ThreadPoolExecutor(max_workers=10)
    futures = []
    for _ in range(100):
        future = pool.submit(account.deposit, 1)
        futures.append(future)
    # 关闭线程池
    pool.shutdown()
    for future in futures:
        future.result()
    print(account.balance)


if __name__ == '__main__':
    main()
```

```python

"""修改上面的程序，启动5个线程向账户中存钱，5个线程从账户中取钱，取钱时如果余额不足就暂停线程进行等待。为了达到上述目标，需要对存钱和取钱的线程进行调度，在余额不足时取钱的线程暂停并释放锁，而存钱的线程将钱存入后要通知取钱的线程，使其从暂停状态被唤醒。可以使用threading模块的Condition来实现线程调度，该对象也是基于锁来创建的，代码如下所示："""

"""
多个线程竞争一个资源 - 保护临界资源 - 锁（Lock/RLock）
多个线程竞争多个资源（线程数>资源数） - 信号量（Semaphore）
多个线程的调度 - 暂停线程执行/唤醒等待中的线程 - Condition
"""
from concurrent.futures import ThreadPoolExecutor
from random import randint
from time import sleep

import threading


class Account:
    """银行账户"""

    def __init__(self, balance=0):
        self.balance = balance
        lock = threading.RLock()
        self.condition = threading.Condition(lock)

    def withdraw(self, money):
        """取钱"""
        with self.condition:
            while money > self.balance:
                self.condition.wait()
            new_balance = self.balance - money
            sleep(0.001)
            self.balance = new_balance

    def deposit(self, money):
        """存钱"""
        with self.condition:
            new_balance = self.balance + money
            sleep(0.001)
            self.balance = new_balance
            self.condition.notify_all()


def add_money(account):
    while True:
        money = randint(5, 10)
        account.deposit(money)
        print(threading.current_thread().name,
              ':', money, '====>', account.balance)
        sleep(0.5)


def sub_money(account):
    while True:
        money = randint(10, 30)
        account.withdraw(money)
        print(threading.current_thread().name,
              ':', money, '<====', account.balance)
        sleep(1)


def main():
    account = Account()
    with ThreadPoolExecutor(max_workers=15) as pool:
        for _ in range(5):
            pool.submit(add_money, account)
        for _ in range(10):
            pool.submit(sub_money, account)


if __name__ == '__main__':
    main()
```

```python
"""多进程：多进程可以有效的解决GIL的问题，实现多进程主要的类是Process，其他辅助的类跟threading模块中的类似，进程间共享数据可以使用管道、套接字等，在multiprocessing模块中有一个Queue类，它基于管道和锁机制提供了多个进程共享的队列。下面是官方文档上关于多进程和进程池的一个示例。"""

"""
多进程和进程池的使用
多线程因为GIL的存在不能够发挥CPU的多核特性
对于计算密集型任务应该考虑使用多进程
time python3 example22.py
real    0m11.512s
user    0m39.319s
sys     0m0.169s
使用多进程后实际执行时间为11.512秒，而用户时间39.319秒约为实际执行时间的4倍
这就证明我们的程序通过多进程使用了CPU的多核特性，而且这台计算机配置了4核的CPU
"""
import concurrent.futures
import math

PRIMES = [
    1116281,
    1297337,
    104395303,
    472882027,
    533000389,
    817504243,
    982451653,
    112272535095293,
    112582705942171,
    112272535095293,
    115280095190773,
    115797848077099,
    1099726899285419
] * 5


def is_prime(n):
    """判断素数"""
    if n % 2 == 0:
        return False

    sqrt_n = int(math.floor(math.sqrt(n)))
    for i in range(3, sqrt_n + 1, 2):
        if n % i == 0:
            return False
    return True


def main():
    """主函数"""
    with concurrent.futures.ProcessPoolExecutor() as executor:
        for number, prime in zip(PRIMES, executor.map(is_prime, PRIMES)):
            print('%d is prime: %s' % (number, prime))


if __name__ == '__main__':
    main()
```

```python
"""重点：多线程和多进程的比较。

以下情况需要使用多线程：

程序需要维护许多共享的状态（尤其是可变状态），Python中的列表、字典、集合都是线程安全的，所以使用线程而不是进程维护共享状态的代价相对较小。
程序会花费大量时间在I/O操作上，没有太多并行计算的需求且不需占用太多的内存。
以下情况需要使用多进程：

程序执行计算密集型任务（如：字节码操作、数据处理、科学计算）。
程序的输入可以并行的分成块，并且可以将运算结果合并。
程序在内存使用方面没有任何限制且不强依赖于I/O操作（如：读写文件、套接字等）。
"""


"""
异步处理：从调度程序的任务队列中挑选任务，该调度程序以交叉的形式执行这些任务，我们并不能保证任务将以某种顺序去执行，因为执行顺序取决于队列中的一项任务是否愿意将 CPU 处理时间让位给另一项任务。异步任务通常通过多任务协作处理的方式来实现，由于执行时间和顺序的不确定，因此需要通过回调式编程或者future对象来获取任务执行的结果。Python 3 通过asyncio模块和await和async关键字（在 Python 3.7 中正式被列为关键字）来支持异步处理。"""

"""
异步I/O - async / await
"""
import asyncio


def num_generator(m, n):
    """指定范围的数字生成器"""
    yield from range(m, n + 1)


async def prime_filter(m, n):
    """素数过滤器"""
    primes = []
    for i in num_generator(m, n):
        flag = True
        for j in range(2, int(i ** 0.5 + 1)):
            if i % j == 0:
                flag = False
                break
        if flag:
            print('Prime =>', i)
            primes.append(i)

        await asyncio.sleep(0.001)
    return tuple(primes)


async def square_mapper(m, n):
    """平方映射器"""
    squares = []
    for i in num_generator(m, n):
        print('Square =>', i * i)
        squares.append(i * i)

        await asyncio.sleep(0.001)
    return squares


def main():
    """主函数"""
    loop = asyncio.get_event_loop()
    future = asyncio.gather(prime_filter(2, 100), square_mapper(1, 100))
    future.add_done_callback(lambda x: print(x.result()))
    loop.run_until_complete(future)
    loop.close()


if __name__ == '__main__':
    main()
```

```python
"""说明：上面的代码使用get_event_loop函数获得系统默认的事件循环，通过gather函数可以获得一个future对象，future对象的add_done_callback可以添加执行完成时的回调函数，loop对象的run_until_complete方法可以等待通过future对象获得协程执行结果。"""
```

```python

"""Python 中有一个名为aiohttp的三方库，它提供了异步的 HTTP 客户端和服务器，这个三方库可以跟asyncio模块一起工作，并提供了对Future对象的支持。Python 3.6中引入了async和await来定义异步执行的函数以及创建异步上下文，在 Python 3.7 中它们正式成为了关键字。下面的代码异步的从5个URL中获取页面并通过正则表达式的命名捕获组提取了网站的标题。"""

import asyncio
import re

import aiohttp

PATTERN = re.compile(r'\<title\>(?P<title>.*)\<\/title\>')


async def fetch_page(session, url):
    async with session.get(url, ssl=False) as resp:
        return await resp.text()


async def show_title(url):
    async with aiohttp.ClientSession() as session:
        html = await fetch_page(session, url)
        print(PATTERN.search(html).group('title'))


def main():
    urls = ('https://www.python.org/',
            'https://git-scm.com/',
            'https://www.jd.com/',
            'https://www.taobao.com/',
            'https://www.douban.com/')
    loop = asyncio.get_event_loop()
    cos = [show_title(url) for url in urls]
    loop.run_until_complete(asyncio.wait(cos))
    loop.close()


if __name__ == '__main__':
    main()
# 重点：异步I/O与多进程的比较。

# 当程序不需要真正的并发性或并行性，而是更多的依赖于异步处理和回调时，asyncio就是一种很好的选择。如果程序中有大量的等待与休眠时，也应该考虑asyncio，它很适合编写没有实时数据处理需求的 Web 应用服务器。
```
