```python
# 排序算法（选择、冒泡和归并）和查找算法（顺序和折半）
def select_sort(items, comp=lambda x, y: x < y):
    """简单选择排序"""
    # items = items[:] 是 Python 中的列表切片复制写法。
    # 等价于 items.copy() 或 list(items)
    items = items[:]
    # 选择排序，每次选择一个最小的元素，与第一个元素交换位置

    for i in range(len(items) - 1):
        min_index = i
        for j in range(i + 1, len(items)):
            if comp(items[j], items[min_index]):
                min_index = j
        items[i], items[min_index] = items[min_index], items[i]
    return items


def bubble_sort(items, comp=lambda x, y: x > y):
    """冒泡排序"""
    items = items[:]
    for i in range(len(items) - 1):
        swapped = False
        for j in range(len(items) - 1 - i):
            if comp(items[j], items[j + 1]):
                items[j], items[j + 1] = items[j + 1], items[j]
                swapped = True
        if not swapped:
            break
    return items


def bubble_sort(items, comp=lambda x, y: x > y):
    """搅拌排序(冒泡排序升级版)"""
    items = items[:]
    for i in range(len(items) - 1):
        swapped = False
        for j in range(len(items) - 1 - i):
            if comp(items[j], items[j + 1]):
                items[j], items[j + 1] = items[j + 1], items[j]
                swapped = True
        if swapped:
            swapped = False
            for j in range(len(items) - 2 - i, i, -1):
                if comp(items[j - 1], items[j]):
                    items[j], items[j - 1] = items[j - 1], items[j]
                    swapped = True
        if not swapped:
            break
    return items


def merge(items1, items2, comp=lambda x, y: x < y):
    """合并(将两个有序的列表合并成一个有序的列表)"""
    items = []
    index1, index2 = 0, 0
    while index1 < len(items1) and index2 < len(items2):
        if comp(items1[index1], items2[index2]):
            items.append(items1[index1])
            index1 += 1
        else:
            items.append(items2[index2])
            index2 += 1

    # 切片 items1[index1:] — 切尾巴
    # 表示：从 index1 位置开始，一直取到列表末尾
    # items1[index1:len(items1)]   # 完整写法
    # items1[index1:]              # 省略 end，默认到末尾

    items += items1[index1:]
    items += items2[index2:]
    return items

# 专业级代码组织方式
# 入口函数
def merge_sort(items, comp=lambda x, y: x < y):
    #list(items) 是 Python 中的列表切片复制写法。
    # 等价于 items.copy() 或 list(items)
    # 设计上保持纯函数风格——不修改输入，只返回新结果。
    return _merge_sort(list(items), comp)

#内部递归函数
# def merge_sort(...):      # 公开接口 → 用户直接调用的
# def _merge_sort(...):     # 私有实现 → 内部使用的
"""
前缀	                含义	                        类比
无前缀	        public —— 公共 API	            餐厅的门面
_ 单下划线	    protected/internal —— 内部使用	厨房内部
__ 双下划线	    private —— 强制名称改写	        后厨密室
"""
def _merge_sort(items, comp):
    """归并排序"""
    if len(items) < 2:
        return items
    mid = len(items) // 2
    # 先通过 _merge_sort() 拆成一个个单元素列表,然后依次通过 merge() 两两进行结合成一个有序列表
    left = _merge_sort(items[:mid], comp)
    right = _merge_sort(items[mid:], comp)
    return merge(left, right, comp)
# 这种设计模式叫做 Facade Pattern（门面模式） 或 Wrapper Function（包装函数）。


def seq_search(items, key):
    """顺序查找"""
    # enumerate 更 Pythonic、更优雅！
    for index, item in enumerate(items):
        if item == key:
            return index
    return -1


def bin_search(items, key):
    # 条件为有序序列
    """折半查找"""
    start, end = 0, len(items) - 1
    # 若 key 不存在 则start > end
    while start <= end:
        mid = (start + end) // 2
        if key > items[mid]:
            start = mid + 1
        elif key < items[mid]:
            end = mid - 1
        else:
            return mid
    return -1
```

```python
"""常用算法：

穷举法 - 又称为暴力破解法，对所有的可能性进行验证，直到找到正确答案。
贪婪法 - 在对问题求解时，总是做出在当前看来最好的选择，不追求最优解，快速找到满意解。
分治法 - 把一个复杂的问题分成两个或更多的相同或相似的子问题，再把子问题分成更小
        的子问题，直到可以直接求解的程度，最后将子问题的解进行合并得到原问题的解
回溯法 - 回溯法又称为试探法，按选优条件向前搜索，当搜索到某一步发现原先选择并不
        优或达不到目标时，就退回一步重新选择。
动态规划 - 基本思想也是将待求解问题分解成若干个子问题，先求解并保存这些子问题的
        解，避免产生大量的重复运算。"""
```

```python
"""贪婪法
"""
"""
贪婪法：在对问题求解时，总是做出在当前看来是最好的选择，不追求最优解，快速找到满意解。
输入：
20 6
电脑      200     20
收音机     20      4
钟       175     10
花瓶      50      2
书       10      1
油画      90      9
"""
class Thing(object):
    """物品"""
    # 构造函数，在创建对象时，自动调用
    def __init__(self, name, price, weight):
        self.name = name
        self.price = price
        self.weight = weight
#   @property  # 将方法变成属性调用
    @property
    def value(self):
        """价格重量比"""
        return self.price / self.weight


def input_thing():
    """输入物品信息"""
    # input得到字符串，split进行分割，得到列表，然后进行解包
    name_str, price_str, weight_str = input().split()
    # 返回值为元组
    return name_str, int(price_str), int(weight_str)


def main():
    """主函数"""

    # map(函数, 可迭代对象)
    # 作用：将可迭代对象的每一个元素依次传入函数，返回结果的迭代器
    # 例：map(lambda x: x * x, [1, 2, 3, 4, 5, 6, 7, 8, 9])
    # 输出：[1, 4, 9, 16, 25, 36, 49, 64, 81]
    # 等价于列表推导式：  [int(x) for x in input().split()]
    max_weight, num_of_things = map(int, input().split())
    all_things = []
    for _ in range(num_of_things):
#         第1步: input_thing() 返回元组
#        ('电脑', 200, 20)
#
#       第2步: * 操作符将元组 "解包" 为独立的参数
#        *('电脑', 200, 20)
#
#        等价于:
#        Thing('电脑', 200, 20)
#
#       第3步: 调用 Thing 的 __init__
#        → 创建一个 Thing 对象
        all_things.append(Thing(*input_thing()))

        # list.sort(key=排序依据函数, reverse=True/False)
        # key	用什么来比较	lambda x: x.value
        # reverse	是否降序	True=从大到小
    all_things.sort(key=lambda x: x.value, reverse=True)
    total_weight = 0
    total_price = 0
    for thing in all_things:
        if total_weight + thing.weight <= max_weight:
            print(f'小偷拿走了{thing.name}')
            total_weight += thing.weight
            total_price += thing.price
    print(f'总价值: {total_price}美元')


# if __name__ == '__main__':                     ║
# ║       main()                                    ║
# ║                                                  ║
# ║   ─────────────────────────────────────          ║
# ║   本质: 一个"开关"                               ║
# ║                                                  ║
# ║   直接运行 → 开关 ON  → 执行 main()             ║
# ║   被导入   → 开关 OFF → 跳过,只提供定义          ║
# ║                                                  ║
# ║   目的:
# ║   避免"导入时就触发副作用"
# ║   ✓ 让同一个文件既能运行也能被导入               ║
# ║   ✓ 防止"导入时就触发副作用"(如弹窗/等待输入等)  ║
# ║   ✓ 是 Python 社区的标准惯例(PEP推荐)           ║
# ║                                                  ║
# ║   记忆口诀:                                     ║
# ║   "如果我是主角(__main__),就开始表演(main())"
if __name__ == '__main__':
    main()
"""
所有语法点汇总表
语法	                        所在行	         含义
class Thing(object)	        333	             定义类，继承自 object
def __init__(self, ...)	    336	             构造方法，初始化对象属性
self.name = name	        337	             绑定实例属性
@property	                341	            装饰器，让方法变属性
*input_thing()	            365	            解包操作符，元组→多参数
lambda x: x.value	        366	            匿名函数，定义排序键
.sort(key=..., reverse=True)366	            列表原地排序
f-string f'{var}'	        371,374	        格式化字符串
if __name__ == '__main__':	377	            程序入口保护
"""
```

```python
# 分治法例子
"""
快速排序 - 选择枢轴对元素进行划分，左边都比枢轴小右边都比枢轴大
"""
def quick_sort(items, comp=lambda x, y: x <= y):
    items = list(items)[:]
    _quick_sort(items, 0, len(items) - 1, comp)
    return items


def _quick_sort(items, start, end, comp):
    if start < end:
        pos = _partition(items, start, end, comp)
        _quick_sort(items, start, pos - 1, comp)
        _quick_sort(items, pos + 1, end, comp)


def _partition(items, start, end, comp):
    pivot = items[end]
    i = start - 1
    for j in range(start, end):
        if comp(items[j], pivot):
            i += 1
            items[i], items[j] = items[j], items[i]
    items[i + 1], items[end] = items[end], items[i + 1]
    return i + 1
```

```python
# 回溯法例子：骑士巡逻。

"""
递归回溯法：叫称为试探法，按选优条件向前搜索，当搜索到某一步，发现原先选择并不优或达不到目标时，就退回一步重新选择，比较经典的问题包括骑士巡逻、八皇后和迷宫寻路等。
"""
import sys
import time

SIZE = 5
total = 0

# 打印棋盘
def print_board(board):
    for row in board:
        for col in row:
            print(str(col).center(4), end='')
        print()


def patrol(board, row, col, step=1):
    if row >= 0 and row < SIZE and \
        col >= 0 and col < SIZE and \
        board[row][col] == 0:
        board[row][col] = step
        if step == SIZE * SIZE:
            global total
            total += 1
            print(f'第{total}种走法: ')
            print_board(board)
        patrol(board, row - 2, col - 1, step + 1)
        patrol(board, row - 1, col - 2, step + 1)
        patrol(board, row + 1, col - 2, step + 1)
        patrol(board, row + 2, col - 1, step + 1)
        patrol(board, row + 2, col + 1, step + 1)
        patrol(board, row + 1, col + 2, step + 1)
        patrol(board, row - 1, col + 2, step + 1)
        patrol(board, row - 2, col + 1, step + 1)
        board[row][col] = 0


def main():
    board = [[0] * SIZE for _ in range(SIZE)]
    patrol(board, SIZE - 1, SIZE - 1)


if __name__ == '__main__':
    main()
```

```python
"""动态规划例子：子列表元素之和的最大值。

说明：子列表指的是列表中索引（下标）连续的元素构成的列表；列表中的元素是int类型，可能包含正整数、0、负整数；程序输入列表中的元素，输出子列表元素求和的最大值，例如：

输入：1 -2 3 5 -3 2

输出：8

输入：0 -2 3 5 -1 2

输出：9

输入：-9 -2 -3 -5 -3

输出：-2"""

def main():
    items = list(map(int, input().split()))
    overall = partial = items[0]
    for i in range(1, len(items)):
        partial = max(items[i], partial + items[i])
        overall = max(partial, overall)
    print(overall)


if __name__ == '__main__':
    main()
# 说明：这个题目最容易想到的解法是使用二重循环，但是代码的时间性能将会变得非常的糟糕。使用动态规划的思想，仅仅是多用了两
# 个变量，就将原来$O(N^2)$复杂度的问题变成了$O(N)$。
```

```python
"""
装饰器函数
使用和取消装饰器"""

# 功能：这是一个计时装饰器，可以给任意函数"装上"计时功能，自动打印该函数的执行耗时
from time import time
# 外层装饰器工厂,实现函数调用的时间记录
def record_time(func):
    """装饰器函数，记录函数执行时间"""
    # 内层
    # *args — 收集所有位置参数 (1, 2, 3)
    # **kwargs — 收集所有关键字参数 {name="张三", age=20}
    def wrapper(*args, **kwargs):
        start = time()
        # 记录原函数的返回值结果
        result = func(*args, **kwargs)
        # 计算并打印函数执行时间
        print(f'{func.__name__}:{time() - start}秒')
        # 返回原函数的返回值结果
        return result
    return wrapper
```
