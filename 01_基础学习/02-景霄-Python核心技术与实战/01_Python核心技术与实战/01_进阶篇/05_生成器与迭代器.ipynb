# 生成器与迭代器（Generator & Iterator）##  学习目标- 理解容器、可迭代对象、迭代器的关系- 掌握生成器（generator）的原理和用法- 能写出自定义生成器- 理解生成器 vs 迭代器的区别- 能用生成器解决实际问题---## 1️⃣ 容器与可迭代对象### 容器（Container）**容器**：包含多个元素的对象。| 容器类型 | 示例 | 底层实现 ||----------|------|----------|| 列表（list） | `[0, 1, 2]` | 动态数组 || 元组（tuple） | `(0, 1, 2)` | 不可变数组 || 字典（dict） | `{0:0, 1:1}` | 哈希表 || 集合（set） | `{0, 1, 2}` | 哈希表 |### 可迭代对象（Iterable）**可迭代对象**：可以用 `for...in...` 遍历的对象。**判断方法**：```pythondef is_iterable(param):try:iter(param)return Trueexcept TypeError:return False```### 示例：判断可迭代对象

````python
# 示例：判断可迭代对象def is_iterable(param):try:iter(param)return Trueexcept TypeError:return Falseparams = [1234,       # 数字'1234',    # 字符串[1, 2, 3, 4],  # 列表{1, 2, 3, 4},  # 集合{1:1, 2:2},    # 字典(1, 2, 3)       # 元组]for param in params:print(f'{param} is iterable? {is_iterable(param)}')
````

---## 2️⃣ 迭代器（Iterator）### 核心概念**迭代器**：- 提供了 `__next__()` 方法- 每次调用返回下一个元素- 没有更多元素时，抛出 `StopIteration` 异常### 示例：自定义迭代器

````python
# 示例：自定义迭代器class MyIterator:def __init__(self, max_num):self.max_num = max_numself.current = 0def __iter__(self):return selfdef __next__(self):if self.current < self.max_num:self.current += 1return self.currentelse:raise StopIteration# 使用迭代器my_iter = MyIterator(3)print(next(my_iter))  # 1print(next(my_iter))  # 2print(next(my_iter))  # 3# print(next(my_iter))  # StopIteration 异常# 也可以用 for 循环（自动处理 StopIteration）my_iter = MyIterator(3)for num in my_iter:print(num)
````

---## 3️⃣ 生成器（Generator）### 核心概念**生成器**：懒人版本的迭代器。- **定义**：包含 `yield` 关键字的函数- **特点**：惰性计算，节省内存- **用法**：调用生成器函数，返回生成器对象### 示例：简单生成器

````python
# 示例：简单生成器def my_generator(n):i = 0while i < n:yield i  # 暂停函数，返回 ii += 1# 创建生成器对象gen = my_generator(3)print(type(gen))  # <class 'generator'># 使用 next() 获取下一个值print(next(gen))  # 0print(next(gen))  # 1print(next(gen))  # 2# print(next(gen))  # StopIteration 异常# 也可以用 for 循环gen = my_generator(3)for num in gen:print(num)
````

### `yield` 关键字详解- **暂停函数**：执行到 `yield` 时，函数暂停，返回值- **保持状态**：局部变量不会被清除，下次从暂停处继续- **恢复执行**：下次调用 `next()` 时，从 `yield` 之后继续### 示例：演示 `yield` 的工作原理

````python
# 示例：演示 yield 的工作原理def demo_yield():print('开始执行')yield 1print('继续执行')yield 2print('执行结束')gen = demo_yield()print('创建生成器')print('第一次调用 next()')print(next(gen))print('第二次调用 next()')print(next(gen))print('第三次调用 next()')try:print(next(gen))except StopIteration:print('生成器耗尽')
````

---## 4️⃣ 生成器 vs 迭代器| 特性 | 迭代器 | 生成器 ||------|----------|----------|| 定义方式 | 实现 `__iter__()` 和 `__next__()` | 包含 `yield` 的函数 || 代码复杂度 | 较高（需要写类） | 较低（函数即可） || 内存占用 | 取决于实现 | 惰性计算，节省内存 || 使用场景 | 复杂迭代逻辑 | 简单迭代、无限序列 |---## 5️⃣ 生成器的特性### 1. 节省内存**对比**：列表 vs 生成器

````python
# 对比：列表 vs 生成器（内存占用）import sys# 列表（占用大量内存）list_comp = [i for i in range(1000000)]print(f'列表占用内存: {sys.getsizeof(list_comp)} bytes')# 生成器（几乎不占内存）gen_exp = (i for i in range(1000000))print(f'生成器占用内存: {sys.getsizeof(gen_exp)} bytes')
````

### 2. 无限序列生成器可以表示无限序列（因为不需要预先生成所有元素）。### 示例：无限斐波那契数列

````python
# 示例：无限斐波那契数列def fibonacci():a, b = 0, 1while True:yield aa, b = b, a + b# 获取前 10 个斐波那契数fib = fibonacci()for i in range(10):print(next(fib), end=' ')
````

---## 6️⃣ 生成器的实际应用### 应用1：大数据处理**场景**：处理大文件（GB级别），不能一次性读入内存。### 示例：逐行读取大文件

````python
# 示例：逐行读取大文件（生成器版本）def read_large_file(file_path):with open(file_path, 'r') as f:for line in f:yield line.strip()# 使用生成器逐行处理# for line in read_large_file('huge_file.txt'):#     process(line)print('生成器让你可以处理超出内存大小的文件！')
````

### 应用2：Pipeline 模式**场景**：数据需要多步处理（例如：读取 → 过滤 → 转换 → 输出）。### 示例：数据处理 Pipeline

````python
# 示例：数据处理 Pipelinedef read_data(file_path):with open(file_path, 'r') as f:for line in f:yield line.strip()def filter_data(data_stream, keyword):for line in data_stream:if keyword in line:yield linedef transform_data(data_stream):for line in data_stream:yield line.upper()# 使用 Pipeline# pipeline = transform_data(filter_data(read_data('data.txt'), 'error'))# for result in pipeline:#     print(result)print('Pipeline 模式让数据处理更清晰、更节省内存！')
````
