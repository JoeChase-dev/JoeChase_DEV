## 一、生成式（推导式）
### 1. 列表生成式
**优点**：
1. 代码简洁优雅
2. 性能优于普通for循环
**强烈建议使用生成式语法创建列表！**

````python
# 普通方式 vs 列表生成式
# 方法1：普通for循环
items1 = []
for i in range(1, 100):
    if i % 3 == 0 and i % 5 == 0:
        items1.append(i)
print(items1)

# 方法2：列表生成式（推荐）
items2 = [i for i in range(1, 100) if i % 3 == 0 and i % 5 == 0]
print(items2)
````

### 2. 字典生成式
**应用场景**：根据条件过滤并转换字典

````python
"""字典生成式示例"""
prices = {
    "AAPL": 191.88,
    "GOOG": 1186.96,
    "IBM": 149.24,
    "ORCL": 48.44,
}

# 生成新字典：只保留value大于100的键值对
prices2 = {key: value for key, value in prices.items() if value > 100}
print(prices2)
````

### 3. 嵌套列表的坑
**错误写法**：
```python
a = [[None] * 3] * 3  # 错误！三行引用同一个列表
a[0][0] = 100
print(a)  # [[100, None, None], [100, None, None], [100, None, None]]
```
**正确写法**：使用列表生成式创建独立的子列表

````python
"""正确的嵌套列表创建方式"""
names = ["关羽", "张飞", "赵云"]
courses = ["语文", "数学", "英语"]

# 正确：每次生成新的子列表
score_list = [[None] * len(courses) for _ in range(len(names))]
score_list[0][0] = 100
print(score_list)  # 只有第一个元素被修改
````

## 二、heapq模块（堆队列）
### 1. 核心功能
- **堆**：一种特殊的二叉树，父节点总是小于等于子节点（最小堆）
- **应用**：从列表中找出最大或最小的N个元素
- **时间复杂度**：$O(N \log N)$

````python
"""heapq模块示例"""
import heapq

# 示例1：找出最大/最小的N个元素
list1 = [34, 25, 18, 87, 76, 95, 48, 65, 28, 14]
print("最大的3个：", heapq.nlargest(3, list1))
print("最小的3个：", heapq.nsmallest(3, list1))

# 示例2：对复杂数据结构排序
list2 = [
    {"name": "IBM", "shares": 100, "price": 91.1},
    {"name": "AAPL", "shares": 50, "price": 543.22},
    {"name": "FB", "shares": 200, "price": 21.09},
]
# key参数：指定排序依据
print("最贵的2个：", heapq.nlargest(2, list2, key=lambda x: x["price"]))
print("最便宜的2个：", heapq.nsmallest(2, list2, key=lambda x: x["price"]))
````

## 三、itertools模块（迭代工具）
### 常用函数
- `permutations()`：全排列（有序）
- `combinations()`：组合（无序）
- `product()`：笛卡尔积（所有组合）
- `cycle()`：无限循环迭代器

````python
"""itertools模块示例"""
import itertools

# 1. 全排列
print("ABCD的全排列（取2个）：")
for perm in itertools.permutations("ABCD", 2):
    print(perm, end=" ")
print()

# 2. 组合
print("ABCDE的组合（取2个）：")
for comb in itertools.combinations("ABCDE", 2):
    print(comb, end=" ")
print()

# 3. 笛卡尔积
print("ABCD和123的笛卡尔积：")
for prod in itertools.product("ABCD", "123"):
    print(prod, end=" ")
print()

# 4. 无限循环（需要break退出）
counter = 0
for item in itertools.cycle(["A", "B", "C"]):
    print(item, end=" ")
    counter += 1
    if counter >= 10:
        break
````

## 四、collections模块（集合工具）
### 常用工具类
- **namedtuple**：命名元组，创建类工厂
- **deque**：双端队列，链表实现，头尾操作$O(1)$
- **Counter**：计数器，dict的子类
- **OrderedDict**：记录插入顺序的字典
- **defaultdict**：带默认工厂函数的字典

````python
"""collections模块示例"""
from collections import Counter, deque, namedtuple

# 1. Counter：统计元素出现次数
words = ["hello", "world", "hello", "python", "hello", "world"]
counter = Counter(words)
print("出现最多的2个单词：", counter.most_common(2))

# 2. deque：双端队列
d = deque([1, 2, 3])
d.appendleft(0)  # 左侧添加
d.append(4)       # 右侧添加
print("deque：", d)
print("左侧弹出：", d.popleft())

# 3. namedtuple：命名元组
Point = namedtuple("Point", ["x", "y"])
p = Point(3, 4)
print(f"坐标：{p.x}, {p.y}")
````

## 五、练习与自测
### 练习题1：使用生成式处理数据
**题目**：给定一个字典，生成一个新字典，只包含值大于25的键值对。
**提示**：使用字典生成式。

````python
"""练习题1解答"""
data = {"A": 10, "B": 20, "C": 30, "D": 40}
result = {k: v for k, v in data.items() if v > 25}
print(result)  # {"C": 30, "D": 40}
````

### 练习题2：使用Counter统计字符频率
**题目**：给定一个字符串，统计每个字符出现的次数，并找出出现次数最多的3个字符。

````python
"""练习题2解答"""
from collections import Counter

def top_3_chars(s):
    counter = Counter(s)
    return counter.most_common(3)

# 测试
s = "abracadabra"
print(top_3_chars(s))
````
