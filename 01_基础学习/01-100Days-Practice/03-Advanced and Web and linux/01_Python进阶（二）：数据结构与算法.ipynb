## 二、排序算法
### 1. 选择排序

```python
"""选择排序"""
def select_sort(items, comp=lambda x, y: x < y):
    """简单选择排序"""
    items = items[:]  # 复制列表，不修改原列表
    for i in range(len(items) - 1):
        min_index = i
        for j in range(i + 1, len(items)):
            if comp(items[j], items[min_index]):
                min_index = j
        items[i], items[min_index] = items[min_index], items[i]
    return items

# 测试
print(select_sort([3, 1, 4, 1, 5, 9, 2, 6]))
```

### 2. 冒泡排序

```python
"""冒泡排序"""
def bubble_sort(items, comp=lambda x, y: x > y):
    """冒泡排序"""
    items = items[:]
    for i in range(len(items) - 1):
        swapped = False
        for j in range(len(items) - 1 - i):
            if comp(items[j], items[j + 1]):
                items[j], items[j + 1] = items[j + 1], items[j]
                swapped = True
        if not swapped:
            break  # 提前退出
    return items

# 测试
print(bubble_sort([3, 1, 4, 1, 5, 9, 2, 6]))
```

### 3. 归并排序

```python
"""归并排序"""
def merge(items1, items2, comp=lambda x, y: x < y):
    """合并两个有序列表"""
    items = []
    index1, index2 = 0, 0
    while index1 < len(items1) and index2 < len(items2):
        if comp(items1[index1], items2[index2]):
            items.append(items1[index1])
            index1 += 1
        else:
            items.append(items2[index2])
            index2 += 1
    items += items1[index1:] + items2[index2:]
    return items

def merge_sort(items, comp=lambda x, y: x < y):
    """归并排序"""
    if len(items) < 2:
        return items[:]
    mid = len(items) // 2
    left = merge_sort(items[:mid], comp)
    right = merge_sort(items[mid:], comp)
    return merge(left, right, comp)

# 测试
print(merge_sort([3, 1, 4, 1, 5, 9, 2, 6]))
```

### 4. 快速排序

```python
"""快速排序"""
def quick_sort(items, comp=lambda x, y: x <= y):
    """快速排序"""
    items = list(items)[:]
    _quick_sort(items, 0, len(items) - 1, comp)
    return items

def _quick_sort(items, start, end, comp):
    if start < end:
        pos = _partition(items, start, end, comp)
        _quick_sort(items, start, pos - 1, comp)
        _quick_sort(items, pos + 1, end, comp)

def _partition(items, start, end, comp):
    pivot = items[end]
    i = start - 1
    for j in range(start, end):
        if comp(items[j], pivot):
            i += 1
            items[i], items[j] = items[j], items[i]
    items[i + 1], items[end] = items[end], items[i + 1]
    return i + 1

# 测试
print(quick_sort([3, 1, 4, 1, 5, 9, 2, 6]))
```

## 三、查找算法
### 1. 顺序查找

```python
"""顺序查找"""
def seq_search(items, key):
    """顺序查找"""
    for index, item in enumerate(items):
        if item == key:
            return index
    return -1

# 测试
print(seq_search([3, 1, 4, 1, 5], 4))  # 2
print(seq_search([3, 1, 4, 1, 5], 6))  # -1
```

### 2. 折半查找（二分查找）

```python
"""折半查找"""
def bin_search(items, key):
    """折半查找（要求有序序列）"""
    start, end = 0, len(items) - 1
    while start <= end:
        mid = (start + end) // 2
        if key > items[mid]:
            start = mid + 1
        elif key < items[mid]:
            end = mid - 1
        else:
            return mid
    return -1

# 测试（有序序列）
print(bin_search([1, 2, 3, 4, 5], 4))  # 3
print(bin_search([1, 2, 3, 4, 5], 6))  # -1
```

## 四、常用算法思想
### 1. 穷举法（暴力破解法）
对所有的可能性进行验证，直到找到正确答案。

```python
"""穷举法示例：百钱百鸡"""
def hundred_chicken():
    """百钱百鸡：公鸡5元/只，母鸡3元/只，小鸡1元/3只"""
    for x in range(0, 21):  # 公鸡最多20只
        for y in range(0, 34):  # 母鸡最多33只
            z = 100 - x - y
            if 5 * x + 3 * y + z // 3 == 100 and z % 3 == 0:
                print(f"公鸡：{x}只，母鸡：{y}只，小鸡：{z}只")


undred_chicken()
```

### 2. 贪婪法
在对问题求解时，总是做出在当前看来最好的选择，不追求最优解，快速找到满意解。

```python
"""贪婪法示例：背包问题"""
class Thing:
    def __init__(self, name, price, weight):
        self.name = name
        self.price = price
        self.weight = weight
    
    @property
    def value(self):
        """价格重量比"""
        return self.price / self.weight

def knapsack(max_weight, things):
    """贪婪法解决背包问题"""
    things.sort(key=lambda x: x.value, reverse=True)
    total_weight = 0
    total_price = 0
    for thing in things:
        if total_weight + thing.weight <= max_weight:
            print(f"拿走了{thing.name}")
            total_weight += thing.weight
            total_price += thing.price
    print(f"总价值: {total_price}")

# 测试
things = [
    Thing("电脑", 200, 20),
    Thing("收音机", 20, 4),
    Thing("钟", 175, 10)
]
knapsack(20, things)
```

### 3. 分治法
把一个复杂的问题分成两个或更多的相同或相似的子问题，再把子问题分成更小的子问题，直到可以直接求解的程度，最后将子问题的解进行合并得到原问题的解。

### 4. 回溯法
又称为试探法，按选优条件向前搜索，当搜索到某一步发现原先选择并不优或达不到目标时，就退回一步重新选择。

```python
"""回溯法示例：骑士巡逻（简化版）"""
def knight_patrol(SIZE=5):
    """骑士巡逻问题"""
    board = [[0] * SIZE for _ in range(SIZE)]
    moves = [
        (-2, -1), (-1, -2), (1, -2), (2, -1),
        (2, 1), (1, 2), (-1, 2), (-2, 1)
    ]
    
    def is_valid(x, y):
        return 0 <= x < SIZE and 0 <= y < SIZE and board[x][y] == 0
    
    def patrol(x, y, step):
        if step > SIZE * SIZE:
            return True
        board[x][y] = step
        for dx, dy in moves:
            nx, ny = x + dx, y + dy
            if is_valid(nx, ny) and patrol(nx, ny, step + 1):
                return True
        board[x][y] = 0
        return False
    
    if patrol(0, 0, 1):
        for row in board:
            print(row)
    else:
        print("未找到解")

# 测试（5x5棋盘）
# knight_patrol(5)
```

### 5. 动态规划
基本思想也是将待求解问题分解成若干个子问题，先求解并保存这些子问题的解，避免产生大量的重复运算。

```python
"""动态规划示例：子列表元素之和的最大值"""
def max_subarray_sum(items):
    """求子列表元素之和的最大值"""
    overall = partial = items[0]
    for i in range(1, len(items)):
        partial = max(items[i], partial + items[i])
        overall = max(partial, overall)
    return overall

# 测试
print(max_subarray_sum([1, -2, 3, 5, -3, 2]))  # 8
print(max_subarray_sum([0, -2, 3, 5, -1, 2]))  # 9
print(max_subarray_sum([-9, -2, -3, -5, -3]))  # -2
```

## 五、练习与自测
### 练习题1：实现插入排序
**题目**：实现插入排序算法，并分析其时间复杂度。
**提示**：从第二个元素开始，将每个元素插入到前面已排序序列的合适位置。

```python
"""练习题1解答：插入排序"""
def insertion_sort(items, comp=lambda x, y: x < y):
    """插入排序"""
    items = items[:]
    for i in range(1, len(items)):
        key = items[i]
        j = i - 1
        while j >= 0 and comp(key, items[j]):
            items[j + 1] = items[j]
            j -= 1
        items[j + 1] = key
    return items

# 测试
print(insertion_sort([3, 1, 4, 1, 5, 9, 2, 6]))
```

### 练习题2：二分查找的递归实现
**题目**：使用递归方式实现二分查找。
**提示**：递归终止条件是`start > end`。

```python
"""练习题2解答：二分查找（递归版）"""
def bin_search_recursive(items, key, start=0, end=None):
    """二分查找（递归版）"""
    if end is None:
        end = len(items) - 1
    if start > end:
        return -1
    mid = (start + end) // 2
    if key > items[mid]:
        return bin_search_recursive(items, key, mid + 1, end)
    elif key < items[mid]:
        return bin_search_recursive(items, key, start, mid - 1)
    else:
        return mid

# 测试
print(bin_search_recursive([1, 2, 3, 4, 5], 4))  # 3
print(bin_search_recursive([1, 2, 3, 4, 5], 6))  # -1
```

## 六、知识总结
### 重点记忆
1. **排序算法**
   - 选择排序：$O(N^2)$，不稳定
   - 冒泡排序：$O(N^2)$，稳定，可提前退出
   - 插入排序：$O(N^2)$，稳定，部分有序时效率高
   - 归并排序：$O(N \log N)$，稳定，需要额外空间
   - 快速排序：平均$O(N \log N)$，不稳定，原地排序
2. **查找算法**
   - 顺序查找：$O(N)$，适用于无序序列
   - 折半查找：$O(\log N)$，适用于有序序列
3. **算法思想**
   - 穷举法：遍历所有可能性
   - 贪婪法：局部最优，不保证全局最优
   - 分治法：分解→解决→合并
   - 回溯法：试探→回退
   - 动态规划：保存子问题解，避免重复计算
