## 一、列表（List）
### 1. 列表的特点
- 有序、可变、允许重复元素
- 使用方括号 `[]` 创建
- 支持索引和切片访问

```python
# 列表的创建和访问
languages = ["Python", "C++", "Java", "Kotlin"]
print(languages[0])  # 访问第一个元素
print(languages[-1])  # 访问最后一个元素
languages.append("Go")  # 追加元素
print(languages)
```

### 2. 列表生成式
**优点**：
1. 代码简洁优雅
2. 性能优于普通for循环
**强烈建议使用生成式语法创建列表！**

```python
# 普通方式 vs 列表生成式
# 方法1：普通for循环
items1 = []
for i in range(1, 100):
    if i % 3 == 0 and i % 5 == 0:
        items1.append(i)
print(items1)

# 方法2：列表生成式（推荐）
items2 = [i for i in range(1, 100) if i % 3 == 0 and i % 5 == 0]
print(items2)
```

## 二、元组（Tuple）
### 1. 元组的特点
- 有序、不可变、允许重复元素
- 使用圆括号 `()` 创建
- 适合存储异构数据（类似记录）
- 可作为字典的键（因为不可变）

```python
# 元组的创建和使用
point = (3, 4)  # 坐标点
name, age = ("张三", 20)  # 元组解包
print(f"名字：{name}，年龄：{age}")

# 元组不可变
try:
    point[0] = 5  # TypeError
except TypeError as e:
    print(f"元组不可修改：{e}")
```

## 三、字符串（String）
### 常用方法
- `strip()`：去除首尾空白
- `split()`：分割字符串
- `join()`：连接字符串
- `find()` / `index()`：查找子串
- `replace()`：替换子串

```python
# 字符串常用操作
text = "  Hello, Python!  "
print(text.strip())  # 去除空白
print(text.lower())  # 转小写
print(text.replace("Python", "World"))  # 替换

# 字符串格式化
name = "张三"
age = 20
print(f"名字：{name}，年龄：{age}")  # f-string（推荐）
```

## 四、练习与自测
### 练习题1：列表去重
**题目**：给定一个列表 `[1, 2, 2, 3, 3, 3, 4, 5, 5]`，去除重复元素，保持原有顺序。
**提示**：使用集合或字典来判重。

```python
"""练习题1解答：列表去重"""
def remove_duplicates(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

# 测试
lst = [1, 2, 2, 3, 3, 3, 4, 5, 5]
print(remove_duplicates(lst))
```

### 练习题2：统计单词频率
**题目**：给定一个字符串，统计每个单词出现的频率。
**示例**：`"hello world hello"` → `{"hello": 2, "world": 1}`

```python
"""练习题2解答：统计单词频率"""
from collections import Counter
def word_count(text):
    words = text.lower().split()
    return dict(Counter(words))

# 测试
text = "hello world hello python world hello"
print(word_count(text))
```
