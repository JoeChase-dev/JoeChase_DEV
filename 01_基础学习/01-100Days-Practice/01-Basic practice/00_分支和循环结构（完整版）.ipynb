## 一、核心概念
### 1. 分支结构（if语句）
Python中使用 `if-elif-else` 进行条件判断：
```python
if condition1:
    # 条件1为真时执行
elif condition2:
    # 条件2为真时执行
else:
    # 其他情况执行
```
### 2. 循环结构
- **for循环**：遍历序列（列表、元组、字符串等）
- **while循环**：条件为真时重复执行
- **break**：跳出整个循环
- **continue**：跳过本次循环，继续下一次

### 示例1：100以内的素数

````python
"""1-100内的素数"""
for i in range(2, 101):
    is_prime = True
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            is_prime = False
            break
    if is_prime:
        print(i, end=" ")
````

### 示例2：斐波那契数列

````python
"""斐波那契数列 - 前20个数"""
a, b = 0, 1
for _ in range(20):
    a, b = b, a + b
    print(a, end=" ")
````

### 示例3：百钱百鸡问题

````python
"""百钱百鸡：公鸡5元/只，母鸡3元/只，小鸡1元/3只，100元买100只鸡"""
# 方法：穷举法
for x in range(0, 21):  # 公鸡最多20只
    for y in range(0, 34):  # 母鸡最多33只
        z = 100 - x - y
        if 5*x + 3*y + z//3 == 100 and z % 3 == 0:
            print(f"公鸡：{x}只，母鸡：{y}只，小鸡：{z}只")
````

## 二、常见错误
### 错误1：缩进错误
```python
# 错误示例
if x > 0:
print('正数')  # IndentationError
```
**修正**：Python使用缩进表示代码块，必须统一使用4个空格。
### 错误2：无限循环
```python
# 错误示例
while True:
    print('无限循环')  # 没有退出条件
```
**修正**：确保循环有退出条件，或使用break跳出。

## 三、练习与自测
### 练习题1：猜数字游戏
**题目**：计算机随机生成1-100的数字，用户有10次机会猜测。
**要求**：每次猜测后提示太大或太小，使用random模块。

````python
"""练习题1解答：猜数字游戏"""
import random
def guess_number():
    target = random.randint(1, 100)
    print("我想了一个1-100之间的数字，你有10次机会！")
    for i in range(10):
        guess = int(input(f"第{i+1}次猜测："))
        if guess == target:
            print(f"恭喜！答案是{target}")
            return
        elif guess > target:
            print("太大了")
        else:
            print("太小了")
    print(f"很遗憾，答案是{target}")

# guess_number()
````

### 练习题2：九九乘法表
**题目**：使用嵌套循环打印上三角形式的九九乘法表。
**输出格式**：
```
1×1=1
1×2=2  2×2=4
1×3=3  2×3=6  3×3=9
...
```

````python
"""练习题2解答：九九乘法表"""
def multiplication_table():
    for i in range(1, 10):
        for j in range(1, i + 1):
            print(f"{j}×{i}={i*j:2d}", end="  ")
        print()  # 换行

multiplication_table()
````

## 四、知识总结
### 重点记忆
1. **分支结构**
   - if-elif-else 处理多分支
   - 条件表达式：`x if condition else y`
2. **循环结构**
   - for循环：遍历序列
   - while循环：条件循环
   - break：跳出整个循环
   - continue：跳过本次循环
3. **应用场景**
   - 遍历序列 → for循环
   - 未知次数 → while循环
   - 提前退出 → break
   - 跳过本次 → continue
