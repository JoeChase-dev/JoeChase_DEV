## 一、文件的打开和关闭
### 1. 操作模式
| 模式 | 含义 | 说明 |
|------|------|------|
| r    | 只读 | 默认模式，文件不存在则报错 |
| w    | 只写 | 文件存在则覆盖，不存在则创建 |
| a    | 追加 | 在文件末尾追加内容 |
| b    | 二进制 | 配合其他模式使用（如rb、wb）|
| +    | 读写 | 配合其他模式使用（如r+）|

### 示例1：读取文本文件

````python
"""读取文本文件的三种方式"""
# 方法1：read() - 读取全部内容
file = open("example.txt", "r", encoding="utf-8")
content = file.read()
print(content)
file.close()

# 方法2：逐行读取
file = open("example.txt", "r", encoding="utf-8")
for line in file:
    print(line, end="")
file.close()

# 方法3：readlines() - 读取为列表
file = open("example.txt", "r", encoding="utf-8")
lines = file.readlines()
print(lines)
file.close()
````

### 示例2：写入文件

````python
"""写入文件"""
# 覆盖写入（w模式）
file = open("output.txt", "w", encoding="utf-8")
file.write("Hello, World!\n")
file.write("Python is great!\n")
file.close()

# 追加写入（a模式）
file = open("output.txt", "a", encoding="utf-8")
file.write("Appended line.\n")
file.close()
````

## 二、异常处理机制
### 1. try-except-finally结构
**目的**：让代码具有健壮性，处理运行时可能出现的错误。
- **try**：放置可能发生异常的代码
- **except**：捕获并处理异常
- **finally**：无论是否异常都会执行（常用于释放资源）
- **else**：try块成功执行后运行（可选）

````python
"""异常处理示例"""
file = None
try:
    file = open("data.txt", "r", encoding="utf-8")
    content = file.read()
    print(content)
except FileNotFoundError:
    print("文件不存在")
except UnicodeDecodeError:
    print("编码错误")
else:
    print("文件读取成功")  # 无异常时执行
finally:
    if file:
        file.close()
    print("资源已释放")
````

### 2. with语句（上下文管理器）
**优势**：自动管理资源，无需手动close()。
**原理**：实现了`__enter__()`和`__exit__()`方法的对象都可以使用with语句。

````python
"""使用with语句（推荐）"""
try:
    with open("data.txt", "r", encoding="utf-8") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("文件不存在")
except Exception as e:
    print(f"发生错误：{e}")

# 文件会在with块结束后自动关闭
````

### 3. 常见异常类型
- **FileNotFoundError**：文件不存在
- **IOError**：输入输出错误
- **ValueError**：值错误（如int("abc")）
- **IndexError**：索引越界
- **KeyError**：字典键不存在
- **TypeError**：类型错误

### 4. 自定义异常

````python
"""自定义异常类"""
class InputError(ValueError):
    pass

def factorial(n):
    if n < 0:
        raise InputError("不能计算负数的阶乘")
    if n in (0, 1):
        return 1
    return n * factorial(n - 1)

# 测试
try:
    print(factorial(5))
    print(factorial(-1))
except InputError as e:
    print(f"输入错误：{e}")
````

## 三、练习与自测
### 练习题1：复制文件
**题目**：编写程序，将源文件的内容复制到目标文件。支持文本文件和二进制文件。
**提示**：使用读写模式，文本用t，二进制用b。

````python
"""练习题1解答：复制文件"""
def copy_file(src, dst, mode="text"):
    """复制文件
    mode: "text" 或 "binary"
    """
    if mode == "text":
        with open(src, "r", encoding="utf-8") as f1:
            with open(dst, "w", encoding="utf-8") as f2:
                f2.write(f1.read())
    else:  # binary mode
        with open(src, "rb") as f1:
            with open(dst, "wb") as f2:
                f2.write(f1.read())
    print(f"已复制：{src} -> {dst}")

# 测试
# copy_file("source.txt", "dest.txt", "text")
# copy_file("image.jpg", "copy.jpg", "binary")
````

### 练习题2：解析CSV文件
**题目**：读取CSV文件，计算某一列的平均值。
**示例CSV**：
```csv
name,score
Alice,85
Bob,90
Charlie,78
```

````python
"""练习题2解答：解析CSV文件"""
def average_score(filename, column_name):
    """计算CSV文件中某列的平均值"""
    total = 0
    count = 0
    with open(filename, "r", encoding="utf-8") as file:
        lines = file.readlines()
        header = lines[0].strip().split(",")
        col_index = header.index(column_name)
        for line in lines[1:]:
            values = line.strip().split(",")
            total += float(values[col_index])
            count += 1
    return total / count if count > 0 else 0

# 测试（需要先创建score.csv文件）
# print(average_score("score.csv", "score"))
````

## 四、知识总结
### 重点记忆
1. **文件操作三步骤**
   - 打开文件：`open()`
   - 读写操作：`read()` / `write()`
   - 关闭文件：`close()` 或使用 `with` 语句
2. **异常处理**
   - 不要滥用异常来处理正常逻辑
   - 捕获具体的异常类型，而非笼统的`Exception`
   - `finally`块用于释放资源
3. **最佳实践**
   - 优先使用`with`语句
   - 指定正确的编码（推荐utf-8）
   - 处理文件操作时总是使用异常处理
