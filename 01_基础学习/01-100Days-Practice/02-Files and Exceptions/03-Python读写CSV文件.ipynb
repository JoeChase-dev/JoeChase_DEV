```python
"""
CSV（Comma Separated Values）全称逗号分隔值文件是一种简单、通用的文件格式
CSV 文件有以下特点：
    纯文本，使用某种字符集（如 ASCII、Unicode、GB2312等）；
    由一条条的记录组成（典型的是每行一条记录）；
    每条记录被分隔符（如逗号、分号、制表符等）分隔为字段（列）；
    每条记录都有同样的字段序列。
"""
```

```python
"""
csv模块实现了读写CSV文件的功能
write 函数会返回 csvwriter 对象，通过 writerow 或 writerows 函数写入数据
writer 还可以使用dialect,dlelimiter,quotechar,quoting来指定分隔符
"""
import csv
import random
with open('score.csv', 'w') as file:
    writer = csv.writer(file,delimiter='|',quoting=csv.QUOTE_ALL)
    writer.writerow(['姓名','语文','数学','英语'])
    names = ['关羽','张飞','赵云','黄忠','马超']
    for name in names:
        scores = [random.randrange(50,100) for _ in range(3)]
        scores.insert(0,name)
        writer.writerow(scores)
```

```python
"""
从CSV文件中读取数据
"""
import csv
with open('score.csv') as file:
    reader = csv.reader(file,delimiter = '|')
    for data_list in reader:
        print(reader.line_num,end = '/t')
        for elem in data_list:
            print(elem,end = '/t')
        print( )
"""
csvreader 对象做for 循环时，每次会取出一个列表对象，包含行内所有的字段
"""
```
