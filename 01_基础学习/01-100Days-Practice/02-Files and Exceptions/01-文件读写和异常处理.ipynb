```python
"""
打开和关闭文件
内置函数 open 打开文件

通过函数的参数指定  文件名、操作模式、编码
操作模式：
r 只读，默认
w 只写，如果文件存在则覆盖，如果文件不存在则创建
x 写入，如果文件存在则报错
a 追加，如果文件存在则在文件末尾追加，如果文件不存在则创建
b 二进制
t 文本
+ 读写

打开文本文件，可以使用encoding参数指定编码，默认为utf-8
打开二进制文件，不指定编码

open打开成功后会返回一个文件对象，可以使用close方法关闭文件

"""
```

```python
"""读写文本文件"""
file = open("01-致橡树.txt", "r", encoding="utf-8")
print(file.read())
file.close()

# 还可以使用for in 和readlines()方法读取文件
print("------------------------")
file = open("01-致橡树.txt", "r", encoding="utf-8")
for line in file:
    print(line,end="")
file.close()

print('\n')
print("------------------------")
file = open("01-致橡树.txt", "r", encoding="utf-8")
lines = file.readlines()  # readlines()方法会将文件中的每一行作为一个元素存储在列表中
print(lines)
file.close()
```

```python
"""写入内容
使用w 或 a 前者会覆盖原有内容，后者会追加到文件末尾
"""
file = open("01-致橡树.txt", "a", encoding="utf-8")
file.write('\n标题：《致橡树》\n')
file.write('\n作者：舒婷\n')
file.write('\n时间：1977年3月')
file.close()
file = open("01-致橡树.txt", "r", encoding="utf-8")
print(file.read())
file.close()
```

```python
"""异常处理机制
    目的：让代码具有健壮性和容错性，可以使用Python的异常处理机制对可能在运行中发生状况的代码进行适当的处理
    关键字：try、except、finally、raise、assert
"""
"""
将运行出现错误的代码放在try中，在try中可以跟上多个except，分别对应不同的错误类型，
except 后可以加上else代码块，这是try中代码运行成功时 会执行的代码块，不会再次在else中异常捕获
最后使用finally来关闭文件，"总是执行的代码块"，适合做释放外部资源的操作
"""
# file 调用失败时，file 的值为 None，成功则为文件对象
file = None

try:
    file = open("01-致橡树.txt", "r", encoding="utf-8")
    print(file.read())
#except 如果异常已经处理就不会再进入后续的except语句
except FileNotFoundError:
    print('无法打开指定文件')
except LookupError:
    print('指定了未知的编码')
except UnicodeDecodeError:
    print('读取文件时解码错误')
finally:
    # 设置初始值 file = None
    if file:
        file.close()

```

```python
"""
Python中的异常都是继承自BaseException的类,四个直接子类：SystemExit、KeyboardInterrupt、GeneratorExit、Exception
SystemExit 是Python解释器退出的异常，KeyboardInterrupt 是用户中断异常，GeneratorExit 是生成器退出异常，Exception 是常规异常的基类
可以自定义异常类型，应该继承自Exception或其子类
raise 关键字（抛出异常对象）

raise 基本语法
1 raise 异常类
2 raise 异常类('错误信息')
3 raise 异常对象

只有当:
    try 块中抛出了异常
    并且 该异常类型 与 except 中指定的类型匹配
    才会进入 except 块。
异常不匹配就会程序崩溃
"""
class InputError(ValueError):
    pass
def fun(num):
    if num < 0:
        raise InputError('只能计算非负整数的阶乘') # 抛出异常对象，并附带错误信息
    if num in (0,1):
        return 1
    return num * fun(num-1)

flag = True
while flag:
    num = int(input('请输入一个非负整数:'))
    try:
        # 如果fun(num)抛出InputError,进入except代码块
        print(f'{num}的阶乘是{fun(num)}')
        flag = False
    except InputError as e:
        print(e)
```

```python
"""上下文管理器
对于open函数返回的对象，还可以使用   with  上下文管理器语法在文件操作完成后
自动执行文件对象的close方法，这样可以让代码变得更加简单优雅，
不需要再写finally代码块来执行关闭文件释放资源的操作
    只有符合上下文管理器协议（有__enter__和__exit__魔术方法）的对象才能使用这种语法


with：
执行顺序:
    调用 open() 返回文件对象
    调用文件对象的 __enter__() 方法
    将 __enter__() 的返回值赋值给 file
    执行 with 块内的代码
    无论是否发生异常,都调用 __exit__() 方法
    __exit__() 内部执行 file.close()
    """
try :
    with open("01-致橡树.txt", "r", encoding="utf-8") as file:
        print(file.read())
except FileNotFoundError:
    print('无法打开指定文件')
except LookupError:
    print('指定了未知的编码')
except UnicodeDecodeError:
    print('读取文件时解码错误')
```

```python
"""
读写二进制文件
"""
try :
    with open('guido.jpg','rb') as file1:
        data = file1.read()
    with open('吉多.jpg','wb') as file2:
        file2.write(data)
except FileNotFoundError:
    print('无法打开指定文件')
except IOError:
    print('读写文件时发生错误')
print('程序执行结束')
```
